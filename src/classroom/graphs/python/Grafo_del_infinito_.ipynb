```python
# @title  Introducción y motivación

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Introducción y motivación</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --code-bg: rgba(14, 165, 233, 0.08);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --code-bg: rgba(255, 255, 255, 0.08);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1080px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 860px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 9000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .topic-content h3 {
      font-size: 1.18rem;
      color: var(--text-primary);
      margin-top: 1rem;
      margin-bottom: 0.75rem;
      padding-left: 0.8rem;
      border-left: 4px solid var(--accent-primary);
    }

    .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0;
      font-size: 1rem;
      color: var(--text-secondary);
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero,
      .block,
      .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-project-diagram"></i>
        Teoría de grafos y retículas cuadradas
      </div>
      <h1 class="main-title">1. Introducción y motivación</h1>
      <p class="subtitle">
        Del plano discreto de puntos enteros a la formulación rigurosa de un grafo, preparando el paso desde una grilla finita hacia la red infinita sobre \( \mathbb{Z}^2 \).
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Planteamiento general</h2>

      <p>
        La teoría de grafos proporciona un lenguaje natural para describir estructuras formadas por objetos discretos y relaciones entre ellos. En muchos problemas matemáticos, físicos y computacionales, esos objetos pueden representarse mediante puntos del plano, mientras que las relaciones se describen como conexiones entre puntos cercanos. Esta manera de modelar resulta especialmente útil cuando se trabaja con coordenadas enteras, ya que los puntos con coordenadas enteras forman una red regular que permite pasar de una intuición geométrica inmediata a una formulación rigurosa en términos de vértices y aristas.
      </p>

      <p>
        El contexto que interesa aquí es el de la retícula cuadrada del plano, construida a partir de puntos de \( \mathbb{Z}^2 \). Esta estructura puede estudiarse en versiones finitas, cuando se restringe a una región concreta, o en su versión infinita, cuando se consideran todos los puntos enteros del plano. En ambos casos, la idea esencial es la misma: cada punto entero se interpreta como un vértice, y se conecta con sus vecinos inmediatos según una regla de adyacencia bien definida.
      </p>

      <p>
        El ejemplo geométrico de la imagen es importante porque muestra, en un entorno pequeño y visualmente manejable, la misma lógica estructural que luego aparecerá en el grafo infinito. Por ello, antes de introducir la retícula infinita \( \mathbb{Z}^2 \), conviene explicar con claridad cómo se construye un grafo a partir de puntos enteros del plano y de una noción natural de vecindad.
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">1.1. La representación de puntos enteros del plano como vértices</span>
            <span class="topic-subtitle">Cómo pasar de puntos de \( \mathbb{Z}^2 \) a nodos de una red</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>Sea</p>

          <div class="formula-box">
            <div class="label">Definición</div>
            \[
            \mathbb{Z} = \{ \ldots, -2, -1, 0, 1, 2, \ldots \}
            \]
          </div>

          <p>
            el conjunto de los números enteros. Consideremos ahora el producto cartesiano
          </p>

          <div class="formula-box">
            <div class="label">Producto cartesiano</div>
            \[
            \mathbb{Z} \times \mathbb{Z} = \{(x,y) : x \in \mathbb{Z},\; y \in \mathbb{Z}\}.
            \]
          </div>

          <p>
            Este conjunto está formado por todos los pares ordenados de enteros, es decir, por todos los puntos del plano cuyas coordenadas son enteras. Es habitual escribir
          </p>

          <div class="formula-box">
            <div class="label">Notación abreviada</div>
            \[
            \mathbb{Z}^2 = \mathbb{Z} \times \mathbb{Z}.
            \]
          </div>

          <p>
            Geométricamente, \( \mathbb{Z}^2 \) representa la red infinita de puntos distribuidos regularmente en el plano con separación unitaria en las direcciones horizontal y vertical.
          </p>

          <p>
            Cada elemento \( (x,y)\in \mathbb{Z}^2 \) puede verse como un punto del plano. Sin embargo, en teoría de grafos se da un paso conceptual adicional: cada uno de esos puntos se interpreta como un vértice. Esta reinterpretación es fundamental, ya que permite transformar una figura geométrica en una estructura combinatoria.
          </p>

          <div class="example-box">
            <strong>Ejemplo.</strong> Los puntos
            \[
            (0,0),\;(1,0),\;(0,1),\;(2,2)
            \]
            pueden considerarse simplemente posiciones del plano, pero también pueden verse como nodos de una red. Una vez adoptada esta segunda interpretación, deja de importar solo dónde están los puntos y pasa a importar además cómo se conectan entre sí.
          </div>

          <p>
            En el ejemplo finito de la imagen, el conjunto de vértices se define como
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de vértices del caso finito</div>
            \[
            V = \{(x,y)\in \mathbb{Z}^2 : 0 \le x \le 2,\; 0 \le y \le 2\}.
            \]
          </div>

          <p>
            Esta expresión selecciona solo aquellos puntos enteros del plano cuyas coordenadas están entre \(0\) y \(2\). Como \(x\) e \(y\) solo pueden tomar los valores \(0,1,2\), se obtiene el conjunto explícito
          </p>

          <div class="formula-box">
            <div class="label">Listado explícito de vértices</div>
            \[
            V =
            \{
            (0,0),(1,0),(2,0),
            (0,1),(1,1),(2,1),
            (0,2),(1,2),(2,2)
            \}.
            \]
          </div>

          <p>
            Así, el ejemplo contiene exactamente \(9\) vértices. Se trata de una pequeña porción cuadrada de la red de puntos enteros del plano.
          </p>

          <p>
            Esta forma de describir vértices mediante coordenadas tiene varias ventajas. En primer lugar, permite una especificación algebraica exacta del conjunto de nodos. En segundo lugar, conserva la intuición geométrica del dibujo. En tercer lugar, hace posible definir de forma natural qué puntos deben considerarse vecinos. Por ello, la representación de puntos enteros del plano como vértices constituye el fundamento del modelo que luego se ampliará al caso infinito.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">1.2. La adyacencia entre vecinos inmediatos</span>
            <span class="topic-subtitle">La regla que transforma puntos en una estructura conectada</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Una vez fijado el conjunto de vértices, el siguiente ingrediente para definir un grafo es establecer cuáles de esos vértices están conectados. Esa relación se llama adyacencia.
          </p>

          <p>
            En este contexto, la adyacencia se define de manera geométrica: dos vértices serán adyacentes si representan puntos enteros situados a distancia unitaria en el plano. Formalmente, dados dos vértices \( (x,y) \) y \( (x',y') \), se dice que son adyacentes si y solo si
          </p>

          <div class="formula-box">
            <div class="label">Regla de adyacencia</div>
            \[
            (x-x')^2 + (y-y')^2 = 1.
            \]
          </div>

          <p>
            Esta condición equivale a imponer que la distancia euclidiana entre ambos puntos sea exactamente \(1\). Como las coordenadas son enteras, esta igualdad solo puede darse cuando los puntos difieren en una unidad en una sola coordenada y coinciden en la otra. Es decir,
          </p>

          <div class="formula-box">
            <div class="label">Vecinos inmediatos</div>
            \[
            (x',y') \in \{(x+1,y),(x-1,y),(x,y+1),(x,y-1)\}.
            \]
          </div>

          <p>
            Por consiguiente, cada vértice puede conectarse únicamente con sus vecinos inmediatos a izquierda, derecha, arriba y abajo. Las diagonales no se consideran aristas.
          </p>

          <div class="example-box">
            <strong>Ejemplo.</strong>
            \[
            (0,0)\sim(1,0), \qquad (0,0)\sim(0,1),
            \]
            pero
            \[
            (0,0)\not\sim(1,1),
            \]
            porque
            \[
            (0-1)^2 + (0-1)^2 = 2 \ne 1.
            \]
          </div>

          <p>
            Si \(V\) es el conjunto de vértices, entonces el conjunto de aristas puede escribirse como
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de aristas</div>
            \[
            E =
            \Big\{
            \{(x,y),(x',y')\}\subseteq V :
            (x-x')^2 + (y-y')^2 = 1
            \Big\}.
            \]
          </div>

          <p>
            De este modo, el grafo queda definido por
          </p>

          <div class="formula-box">
            <div class="label">Definición del grafo</div>
            \[
            G = (V,E).
            \]
          </div>

          <p>
            Esta regla de adyacencia tiene varias consecuencias importantes. En primer lugar, la relación es simétrica: si un vértice es adyacente a otro, entonces el segundo es adyacente al primero. Por tanto, el grafo es no dirigido. En segundo lugar, la conexión es estrictamente local, lo que da lugar a una estructura de red regular. En tercer lugar, la definición es perfectamente compatible con movimientos discretos de una unidad, lo que hace que este tipo de grafos sea ideal para estudiar caminos y paseos aleatorios.
          </p>

          <p>
            En el ejemplo de la imagen, el vértice central \( (1,1) \) tiene cuatro vecinos:
          </p>

          <div class="formula-box">
            <div class="label">Vecinos del centro</div>
            \[
            (0,1),\;(2,1),\;(1,0),\;(1,2),
            \]
          </div>

          <p>
            mientras que una esquina como \( (0,0) \) solo tiene dos:
          </p>

          <div class="formula-box">
            <div class="label">Vecinos de una esquina</div>
            \[
            (1,0),\;(0,1).
            \]
          </div>

          <p>
            Esto muestra que, en un grafo finito, la presencia de fronteras modifica el número de vecinos posibles. Más adelante, en el caso infinito, esa restricción desaparecerá y cada vértice tendrá exactamente cuatro vecinos.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">1.3. Del ejemplo geométrico al modelo de teoría de grafos</span>
            <span class="topic-subtitle">La traducción desde una figura visible hacia una definición abstracta</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            El verdadero salto conceptual ocurre cuando se entiende que la figura geométrica no es solo una imagen, sino la representación de una estructura abstracta. En efecto, una vez especificados el conjunto de vértices \(V\) y la relación de adyacencia, queda definido un grafo en sentido formal.
          </p>

          <p>
            En el ejemplo considerado, el conjunto de vértices es
          </p>

          <div class="formula-box">
            <div class="label">Vértices del ejemplo</div>
            \[
            V = \{(x,y)\in \mathbb{Z}^2 : 0 \le x \le 2,\; 0 \le y \le 2\},
            \]
          </div>

          <p>
            y la adyacencia viene dada por
          </p>

          <div class="formula-box">
            <div class="label">Relación de adyacencia</div>
            \[
            (x,y)\sim(x',y')
            \iff
            (x-x')^2 + (y-y')^2 = 1.
            \]
          </div>

          <p>
            Por tanto, el grafo asociado es
          </p>

          <div class="formula-box">
            <div class="label">Grafo asociado</div>
            \[
            G = (V,E),
            \]
            donde
            \[
            E = \{\{u,v\}\subseteq V : u\sim v\}.
            \]
          </div>

          <p>
            A partir de este momento, el objeto puede estudiarse con todas las herramientas de la teoría de grafos. El dibujo ya no es esencial; lo esencial es la estructura de conexiones.
          </p>

          <p>
            Esta reformulación es importante por varias razones. Primero, permite separar el objeto matemático de su representación visual. Un mismo grafo puede dibujarse de diferentes maneras, pero sigue siendo el mismo siempre que conserve las mismas relaciones de adyacencia. Segundo, abre la puerta al uso de conceptos formales como grado, camino, distancia, ciclo, subgrafo y conectividad. Tercero, permite generalizar el modelo de forma natural.
          </p>

          <p>
            En efecto, el ejemplo de la imagen corresponde a una región pequeña del plano. Si se eliminan las restricciones
          </p>

          <div class="formula-box">
            <div class="label">Restricciones del caso finito</div>
            \[
            0 \le x \le 2, \qquad 0 \le y \le 2,
            \]
          </div>

          <p>
            y se consideran todos los puntos de \( \mathbb{Z}^2 \), se obtiene el grafo infinito de la red cuadrada. Es decir, se pasa del conjunto finito \(V\) al conjunto infinito
          </p>

          <div class="formula-box">
            <div class="label">Conjunto infinito de vértices</div>
            \[
            V_\infty = \mathbb{Z}^2.
            \]
          </div>

          <p>
            Manteniendo la misma regla de adyacencia, se obtiene el grafo infinito
          </p>

          <div class="formula-box">
            <div class="label">Grafo infinito</div>
            \[
            G_\infty = (V_\infty,E_\infty),
            \]
            donde
            \[
            E_\infty =
            \Big\{
            \{(x,y),(x',y')\}\subseteq \mathbb{Z}^2 :
            (x-x')^2 + (y-y')^2 = 1
            \Big\}.
            \]
          </div>

          <p>
            Así, el ejemplo geométrico finito debe entenderse como una ventana local de una estructura mucho mayor. La teoría de grafos proporciona precisamente el marco formal que permite reconocer esa continuidad entre el caso finito y el infinito.
          </p>

          <p>
            En resumen, el paso del ejemplo geométrico al modelo de teoría de grafos consiste en traducir una configuración visible de puntos y vecindades a una definición abstracta basada en conjuntos y relaciones. Esa traducción no solo aporta rigor, sino que prepara el camino para el estudio posterior de paseos aleatorios, retornos al origen, recurrencia y propiedades asintóticas sobre la grilla infinita \( \mathbb{Z}^2 \).
          </p>
        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        Construcción de un grafo a partir de puntos enteros del plano y una regla de vecindad.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_intro_grafo_acordeon', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_intro_grafo_acordeon') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
# @title

#  Grafo finito

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>El grafo finito de la imagen</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
      --table-header: rgba(14, 165, 233, 0.12);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
      --table-header: rgba(255, 255, 255, 0.08);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 920px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 12000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    .grid-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-radius: 16px;
      padding: 1rem;
      margin: 1rem 0 1.1rem;
      overflow-x: auto;
      box-shadow: var(--shadow-soft);
    }

    .grid-3x3 {
      display: grid;
      grid-template-columns: repeat(3, minmax(120px, 1fr));
      gap: 0.65rem;
      max-width: 520px;
      margin: 0 auto;
    }

    .grid-cell {
      border: 1px solid var(--border-color);
      border-radius: 12px;
      padding: 0.85rem 0.6rem;
      text-align: center;
      background: var(--bg-tertiary);
      color: var(--text-primary);
      font-family: 'JetBrains Mono', monospace;
      font-size: 0.95rem;
      font-weight: 500;
    }

    table {
      width: 100%;
      border-collapse: collapse;
      margin: 1rem 0 1.1rem;
      overflow: hidden;
      border-radius: 14px;
      box-shadow: var(--shadow-soft);
    }

    th, td {
      border: 1px solid var(--border-color);
      padding: 0.85rem 0.9rem;
      text-align: left;
      color: var(--text-secondary);
      vertical-align: top;
    }

    th {
      background: var(--table-header);
      color: var(--text-primary);
      font-weight: 700;
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
      .grid-3x3 {
        grid-template-columns: repeat(3, minmax(90px, 1fr));
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-border-all"></i>
        Grilla finita y estructura discreta
      </div>
      <h1 class="main-title">2. El grafo finito de la imagen</h1>
      <p class="subtitle">
        Una malla cuadrada \(3 \times 3\) que resume, en una estructura pequeña y completamente explícita, la lógica de vecindad local, borde, conectividad y representación algebraica propia de la retícula cuadrada.
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        El grafo finito de la imagen constituye una versión acotada de la retícula cuadrada del plano. Su interés matemático radica en que permite introducir, en una situación completamente explícita, la estructura de vecindad característica de las grillas cartesianas. Aunque el objeto es pequeño, contiene ya todos los elementos esenciales que luego aparecerán en la formulación del grafo infinito: un conjunto de vértices descrito por coordenadas enteras, una regla precisa de adyacencia, una interpretación geométrica como malla y un conjunto de propiedades combinatorias básicas.
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">2.1. Definición del conjunto de vértices</span>
            <span class="topic-subtitle">Qué puntos enteros del plano forman parte del grafo</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            El primer paso consiste en especificar qué puntos del plano forman parte del grafo. En la imagen, los vértices están dados por todos los puntos de coordenadas enteras contenidos en el cuadrado determinado por \(0 \le x \le 2\) y \(0 \le y \le 2\). Por tanto,
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de vértices</div>
            \[
            V=\{(x,y)\in \mathbb{Z}\times\mathbb{Z}:0\le x\le 2,\;0\le y\le 2\}.
            \]
          </div>

          <p>
            Como las únicas posibilidades para cada coordenada son
          </p>

          <div class="formula-box">
            <div class="label">Valores posibles de las coordenadas</div>
            \[
            x\in\{0,1,2\},\qquad y\in\{0,1,2\},
            \]
          </div>

          <p>
            el conjunto de vértices puede escribirse explícitamente como
          </p>

          <div class="formula-box">
            <div class="label">Listado explícito</div>
            \[
            V=
            \{
            (0,0),(1,0),(2,0),
            (0,1),(1,1),(2,1),
            (0,2),(1,2),(2,2)
            \}.
            \]
          </div>

          <p>
            De aquí se deduce inmediatamente que el número de vértices es
          </p>

          <div class="formula-box">
            <div class="label">Cardinalidad</div>
            \[
            |V|=3\cdot 3=9.
            \]
          </div>

          <p>
            Cada uno de estos pares ordenados representa simultáneamente un punto del plano de coordenadas enteras y un vértice del grafo. Esta forma de describir los nodos es especialmente útil porque proporciona una representación algebraica precisa y, al mismo tiempo, conserva la intuición geométrica del dibujo.
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">2.2. Definición de la adyacencia</span>
            <span class="topic-subtitle">Cuándo dos vértices están conectados por una arista</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Una vez determinado el conjunto de vértices, el segundo paso es definir cuándo dos vértices están conectados por una arista. La regla es la misma que se utiliza en la retícula cuadrada: dos vértices son adyacentes si están a distancia \(1\) en el plano.
          </p>

          <p>
            Dados \((x,y)\) y \((x',y')\) en \(V\), se define la adyacencia mediante la condición
          </p>

          <div class="formula-box">
            <div class="label">Relación de adyacencia</div>
            \[
            (x,y)\sim(x',y')
            \iff
            (x-x')^2+(y-y')^2=1.
            \]
          </div>

          <p>
            Esta igualdad expresa que los dos puntos difieren en una unidad en una sola coordenada y coinciden en la otra. Equivalentemente,
          </p>

          <div class="formula-box">
            <div class="label">Vecinos inmediatos</div>
            \[
            (x',y')\in\{(x+1,y),(x-1,y),(x,y+1),(x,y-1)\},
            \]
          </div>

          <p>
            siempre que esos puntos pertenezcan también a \(V\).
          </p>

          <p>
            Por tanto, el conjunto de aristas queda definido como
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de aristas</div>
            \[
            E=
            \Big\{
            \{(x,y),(x',y')\}\subseteq V:
            (x-x')^2+(y-y')^2=1
            \Big\}.
            \]
          </div>

          <p>
            Esta definición excluye las diagonales. Por ejemplo,
          </p>

          <div class="example-box">
            <strong>Ejemplo.</strong>
            \[
            (0,0)\sim(1,0), \qquad (0,0)\sim(0,1),
            \]
            pero
            \[
            (0,0)\not\sim(1,1),
            \]
            porque
            \[
            (0-1)^2+(0-1)^2=2\neq 1.
            \]
          </div>

          <p>
            Si se listan todas las aristas sin repetir, se obtiene:
          </p>

          <div class="formula-box">
            <div class="label">Listado de aristas</div>
            \[
            \begin{aligned}
            E=\{&
            \{(0,0),(1,0)\},\{(1,0),(2,0)\},\\
            &\{(0,1),(1,1)\},\{(1,1),(2,1)\},\\
            &\{(0,2),(1,2)\},\{(1,2),(2,2)\},\\
            &\{(0,0),(0,1)\},\{(0,1),(0,2)\},\\
            &\{(1,0),(1,1)\},\{(1,1),(1,2)\},\\
            &\{(2,0),(2,1)\},\{(2,1),(2,2)\}
            \}.
            \end{aligned}
            \]
          </div>

          <p>
            Por tanto, el número total de aristas es
          </p>

          <div class="formula-box">
            <div class="label">Número total de aristas</div>
            \[
            |E|=12.
            \]
          </div>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">2.3. Interpretación geométrica como malla \(3\times 3\)</span>
            <span class="topic-subtitle">Cómo leer el grafo como una pequeña cuadrícula cartesiana</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Geométricamente, este grafo representa una malla cuadrada \(3\times 3\). Eso significa que los vértices están organizados en tres filas y tres columnas:
          </p>

          <div class="grid-box">
            <div class="grid-3x3">
              <div class="grid-cell">(0,2)</div>
              <div class="grid-cell">(1,2)</div>
              <div class="grid-cell">(2,2)</div>

              <div class="grid-cell">(0,1)</div>
              <div class="grid-cell">(1,1)</div>
              <div class="grid-cell">(2,1)</div>

              <div class="grid-cell">(0,0)</div>
              <div class="grid-cell">(1,0)</div>
              <div class="grid-cell">(2,0)</div>
            </div>
          </div>

          <p>
            Las aristas unen únicamente los vecinos inmediatos en sentido horizontal y vertical. Desde esta perspectiva, el grafo puede entenderse como una pequeña porción de la red cuadrada del plano. Cada vértice ocupa una posición regular, y las conexiones siguen el patrón estándar de una grilla cartesiana.
          </p>

          <p>
            Esta interpretación geométrica es importante por varias razones: hace visible la estructura local del grafo, permite distinguir de forma inmediata esquinas, bordes y centro, anticipa la extensión al caso infinito y facilita la comprensión intuitiva de caminos y recorridos.
          </p>

          <p>
            De hecho, este grafo puede verse como el resultado de recortar una ventana cuadrada de lado \(3\) dentro de la retícula infinita \(\mathbb{Z}^2\). En ese sentido, no es una estructura aislada, sino una versión finita y visualmente manejable de una construcción más general.
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">2.4. Qué hace y qué modela el grafo finito</span>
            <span class="topic-subtitle">La red como estructura de vecindad y como modelo de movimientos discretos</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Desde el punto de vista matemático, el grafo finito cumple una doble función: representa una estructura de vecindad y modela movimientos discretos sobre una región acotada.
          </p>

          <p>
            En primer lugar, el grafo describe formalmente qué puntos del plano están conectados entre sí. Es decir, modela la relación ser vecino inmediato dentro de una pequeña región de la red cuadrada. No se trata ya de un simple dibujo, sino de una estructura combinatoria concreta sobre la que se pueden estudiar propiedades exactas.
          </p>

          <p>
            En segundo lugar, este grafo finito sirve como modelo para distintos procesos discretos. Entre ellos se encuentran recorridos entre nodos, conteo de caminos, difusión restringida a una región, movimientos paso a paso y versiones finitas de paseos aleatorios.
          </p>

          <p>
            Por ejemplo, si una partícula o caminante se encuentra en el vértice \((1,1)\), el grafo describe exactamente cuáles son sus movimientos permitidos:
          </p>

          <div class="formula-box">
            <div class="label">Movimientos desde el centro</div>
            \[
            (1,1)\to(0,1),\qquad
            (1,1)\to(2,1),\qquad
            (1,1)\to(1,0),\qquad
            (1,1)\to(1,2).
            \]
          </div>

          <p>
            En cambio, si el caminante está en la esquina \((0,0)\), solo puede desplazarse a
          </p>

          <div class="formula-box">
            <div class="label">Movimientos desde una esquina</div>
            \[
            (1,0)\quad \text{o} \quad (0,1).
            \]
          </div>

          <p>
            Así, el grafo finito modela una dinámica local condicionada por la presencia de fronteras. Esto es precisamente lo que lo diferencia del grafo infinito: aquí aparecen límites espaciales que restringen el movimiento.
          </p>

          <p>
            Además, el grafo finito tiene un fuerte valor pedagógico. Permite introducir, en un contexto pequeño y controlable, conceptos que luego serán esenciales en el caso infinito: vecindad, aristas, grado, conectividad, caminos y estructura de red.
          </p>

          <p>
            En resumen, este grafo finito no solo muestra una malla \(3\times 3\), sino que formaliza una red de conexiones locales y sirve como modelo elemental para procesos discretos en una grilla acotada.
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">2.5. Grados de vértices: esquinas, bordes y centro</span>
            <span class="topic-subtitle">Cómo cambia el número de vecinos según la posición</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Una propiedad básica de todo grafo es el grado de cada vértice, es decir, el número de vértices adyacentes a él. Si \(v\in V\), su grado se denota por \(\deg(v)\).
          </p>

          <p>
            En este grafo, el grado de un vértice depende de su posición geométrica dentro de la malla. Aparecen tres tipos de vértices.
          </p>

          <div class="example-box">
            <strong>a) Esquinas.</strong> Son los vértices
            \[
            (0,0),\;(2,0),\;(0,2),\;(2,2).
            \]
            Cada uno tiene exactamente dos vecinos. Por ejemplo,
            \[
            N(0,0)=\{(1,0),(0,1)\},
            \qquad \deg(0,0)=2.
            \]
            Análogamente,
            \[
            \deg(2,0)=\deg(0,2)=\deg(2,2)=2.
            \]
          </div>

          <div class="example-box">
            <strong>b) Bordes no esquina.</strong> Son los vértices
            \[
            (1,0),\;(0,1),\;(2,1),\;(1,2).
            \]
            Cada uno tiene tres vecinos. Por ejemplo,
            \[
            N(1,0)=\{(0,0),(2,0),(1,1)\},
            \qquad \deg(1,0)=3.
            \]
            Del mismo modo,
            \[
            \deg(0,1)=\deg(2,1)=\deg(1,2)=3.
            \]
          </div>

          <div class="example-box">
            <strong>c) Centro.</strong> El vértice central es
            \[
            (1,1),
            \]
            y tiene cuatro vecinos:
            \[
            N(1,1)=\{(0,1),(2,1),(1,0),(1,2)\},
            \qquad \deg(1,1)=4.
            \]
          </div>

          <p>
            En consecuencia, la distribución de grados del grafo es:
          </p>

          <table>
            <thead>
              <tr>
                <th>Tipo de vértice</th>
                <th>Grado</th>
                <th>Cantidad</th>
              </tr>
            </thead>
            <tbody>
              <tr>
                <td>Esquina</td>
                <td>2</td>
                <td>4</td>
              </tr>
              <tr>
                <td>Borde no esquina</td>
                <td>3</td>
                <td>4</td>
              </tr>
              <tr>
                <td>Centro</td>
                <td>4</td>
                <td>1</td>
              </tr>
            </tbody>
          </table>

          <p>
            Esto puede resumirse mediante la suma de grados:
          </p>

          <div class="formula-box">
            <div class="label">Suma de grados</div>
            \[
            \sum_{v\in V}\deg(v)=4\cdot 2+4\cdot 3+1\cdot 4=8+12+4=24.
            \]
          </div>

          <p>
            Por el lema del apretón de manos,
          </p>

          <div class="formula-box">
            <div class="label">Lema del apretón de manos</div>
            \[
            \sum_{v\in V}\deg(v)=2|E|,
            \]
            y como \(|E|=12\), se verifica que
            \[
            2|E|=2\cdot 12=24,
            \]
            lo que confirma la consistencia del conteo.
          </div>

          <p>
            Una consecuencia importante es que este grafo no es regular, ya que no todos sus vértices tienen el mismo grado. La falta de regularidad se debe a la presencia del borde. Más adelante, en la grilla infinita, esta irregularidad desaparecerá y todos los vértices tendrán grado \(4\).
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">2.6. Propiedades básicas del grafo finito</span>
            <span class="topic-subtitle">Las características estructurales fundamentales de la malla \(3\times 3\)</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Una vez definidos vértices, aristas y grados, pueden enunciarse las propiedades estructurales más importantes del grafo.
          </p>

          <div class="example-box">
            <strong>a) Es un grafo finito.</strong>
            Tiene un número finito de vértices y aristas, concretamente
            \[
            |V|=9,\qquad |E|=12.
            \]
          </div>

          <div class="example-box">
            <strong>b) Es un grafo simple.</strong>
            No tiene lazos ni aristas múltiples. Ningún vértice está conectado consigo mismo, y entre dos vértices distintos hay a lo sumo una arista.
          </div>

          <div class="example-box">
            <strong>c) Es no dirigido.</strong>
            La adyacencia es simétrica: si \(u\sim v\), entonces \(v\sim u\). Por eso las aristas se escriben como pares no ordenados \(\{u,v\}\).
          </div>

          <div class="example-box">
            <strong>d) Es conexo.</strong>
            Dados dos vértices cualesquiera \(u=(x,y)\) y \(v=(x',y')\), siempre existe un camino entre ellos. En efecto, basta moverse horizontalmente hasta ajustar la primera coordenada y luego verticalmente hasta ajustar la segunda.
          </div>

          <div class="example-box">
            <strong>e) Es bipartito.</strong>
            El conjunto de vértices puede dividirse en dos clases según la paridad de \(x+y\):
            \[
            A=\{(x,y)\in V: x+y \text{ es par}\},
            \qquad
            B=\{(x,y)\in V: x+y \text{ es impar}\}.
            \]
            Toda arista conecta un vértice de \(A\) con uno de \(B\), porque moverse una unidad cambia la paridad de \(x+y\). En este caso,
            \[
            A=\{(0,0),(2,0),(1,1),(0,2),(2,2)\},
            \]
            \[
            B=\{(1,0),(0,1),(2,1),(1,2)\}.
            \]
            Esto implica, entre otras cosas, que el grafo no contiene ciclos de longitud impar.
          </div>

          <div class="example-box">
            <strong>f) Tiene estructura planar evidente.</strong>
            Puede dibujarse en el plano sin cruces de aristas; de hecho, ya está dado de ese modo por su propia construcción geométrica. Por tanto, es un grafo planar.
          </div>

          <div class="example-box">
            <strong>g) Es una red con borde.</strong>
            No todos los vértices tienen el mismo entorno local. Los vértices del interior tienen más vecinos que los del borde, y los del borde más que los de las esquinas. Esta presencia de frontera es una característica esencial del caso finito.
          </div>

          <div class="example-box">
            <strong>h) Es un subgrafo inducido de \(\mathbb{Z}^2\).</strong>
            Si se considera la red infinita \(\mathbb{Z}^2\), este grafo se obtiene restringiendo los vértices al conjunto
            \[
            V=\{(x,y)\in \mathbb{Z}^2:0\le x\le 2,\;0\le y\le 2\}
            \]
            y conservando todas las aristas entre ellos. Por tanto, es un subgrafo inducido de la retícula infinita.
          </div>

          <div class="example-box">
            <strong>i) Admite interpretación como producto cartesiano de caminos.</strong>
            También puede identificarse con
            \[
            G\cong P_3 \square P_3,
            \]
            donde \(P_3\) es el grafo camino con \(3\) vértices. Esta observación lo sitúa dentro de una familia clásica de grafos de malla rectangular.
          </div>

          <p>
            El grafo finito de la imagen es pequeño, pero conceptualmente muy rico. Resume en una estructura concreta varias ideas fundamentales: puntos enteros del plano como vértices, vecindad inmediata como adyacencia, malla geométrica como grafo e interior y frontera como fuentes de distinta conectividad. Por eso este ejemplo no debe verse como un caso menor, sino como el laboratorio ideal para introducir la lógica del modelo general.
          </p>

        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        El grafo finito como malla discreta con frontera.
      </p>
      <p>
        El grafo finito de la imagen es una pequeña malla cartesiana formada por vértices de coordenadas enteras y aristas que unen vecinos inmediatos. Su importancia no reside solo en su sencillez visual, sino en que encarna de manera exacta una estructura de red discreta con frontera.
      </p>
      <p>
        Esa frontera altera el comportamiento local de los vértices: las esquinas tienen menos vecinos que el centro, y los bordes tienen un grado intermedio. Por ello, el grafo no es regular. Sin embargo, sigue conservando propiedades globales importantes, como conectividad, planaridad y bipartición.
      </p>
      <p>
        Desde el punto de vista conceptual, este grafo funciona como un modelo introductorio excelente. Permite entender cómo se construyen redes a partir de coordenadas enteras, cómo se formaliza la adyacencia y cómo la geometría del dibujo se traduce en propiedades combinatorias. Además, sirve como antesala natural del grafo infinito \(\mathbb{Z}^2\), donde las mismas ideas se desarrollan sin frontera y con un alcance matemático mucho más amplio.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_grafo_finito', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_grafo_finito') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))

```

```python
# @title  Del grafo finito al grafo infinito

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Del grafo finito al grafo infinito</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
      --table-header: rgba(14, 165, 233, 0.12);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
      --table-header: rgba(255, 255, 255, 0.08);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 920px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 12000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    .sequence-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-radius: 16px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
      text-align: center;
      font-family: 'JetBrains Mono', monospace;
      color: var(--text-primary);
      font-size: 0.98rem;
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-expand-arrows-alt"></i>
        De la ventana local a la estructura global
      </div>
      <h1 class="main-title">3. Del grafo finito al grafo infinito</h1>
      <p class="subtitle">
        El paso desde la malla finita hacia la red infinita sobre \( \mathbb{Z}^2 \) no introduce una regla nueva: extiende la misma vecindad local a todo el plano discreto y elimina los efectos de frontera.
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        El paso del grafo finito al grafo infinito constituye una transición central en el desarrollo del modelo. El grafo finito de la imagen permite introducir la construcción de una red a partir de puntos enteros y vecindad inmediata, pero su tamaño reducido y la presencia de frontera impiden captar las propiedades globales más profundas. Por ello, es necesario interpretar ese grafo como una porción local de la red cuadrada infinita definida sobre \( \mathbb{Z}^2 \).
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">3.1. El grafo finito como subgrafo inducido</span>
            <span class="topic-subtitle">La malla finita como restricción de la red cuadrada infinita</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            El grafo finito de la imagen se construye a partir del conjunto de vértices
          </p>

          <div class="formula-box">
            <div class="label">Vértices del grafo finito</div>
            \[
            V_{\mathrm{fin}}=
            \{(x,y)\in \mathbb{Z}^2:0\le x\le 2,\;0\le y\le 2\},
            \]
          </div>

          <p>
            y de la regla de adyacencia
          </p>

          <div class="formula-box">
            <div class="label">Regla de vecindad</div>
            \[
            (x,y)\sim(x',y')
            \iff
            (x-x')^2+(y-y')^2=1.
            \]
          </div>

          <p>
            Esta misma regla de vecindad es la que define la red cuadrada sobre todo \( \mathbb{Z}^2 \). Por ello, el grafo finito no introduce una estructura nueva, sino que restringe una estructura más general a una región acotada del plano discreto.
          </p>

          <p>
            Para formalizar esta idea, conviene recordar la noción de subgrafo inducido. Si \(G=(V_G,E_G)\) es un grafo y \(W\subseteq V_G\), entonces el subgrafo inducido por \(W\), denotado \(G[W]\), es el grafo cuyos vértices son los elementos de \(W\) y cuyas aristas son exactamente aquellas aristas de \(G\) cuyos dos extremos pertenecen a \(W\). Es decir,
          </p>

          <div class="formula-box">
            <div class="label">Subgrafo inducido</div>
            \[
            G[W]=\bigl(W,\{\{u,v\}\in E_G:\;u,v\in W\}\bigr).
            \]
          </div>

          <p>
            Ahora bien, si definimos el grafo infinito de la retícula cuadrada como
          </p>

          <div class="formula-box">
            <div class="label">Grafo infinito</div>
            \[
            G_\infty=(V_\infty,E_\infty),
            \qquad
            V_\infty=\mathbb{Z}^2
            \]
            y
            \[
            E_\infty=
            \Bigl\{
            \{(x,y),(x',y')\}\subseteq \mathbb{Z}^2:
            (x-x')^2+(y-y')^2=1
            \Bigr\},
            \]
          </div>

          <p>
            entonces el grafo finito de la imagen se obtiene tomando el subconjunto
          </p>

          <div class="formula-box">
            <div class="label">Recorte finito</div>
            \[
            W=
            \{(x,y)\in \mathbb{Z}^2:0\le x\le 2,\;0\le y\le 2\}
            \]
          </div>

          <p>
            y formando el subgrafo inducido correspondiente:
          </p>

          <div class="formula-box">
            <div class="label">Relación exacta entre ambos grafos</div>
            \[
            G_{\mathrm{fin}}=G_\infty[W].
            \]
          </div>

          <p>
            Esta igualdad tiene un significado importante. Indica que el grafo finito hereda del grafo infinito tanto sus vértices como su regla de conexión. No hay una elección arbitraria de aristas: se conservan exactamente las conexiones que ya existían en la red infinita entre los vértices del cuadrado considerado.
          </p>

          <p>
            Dicho de otra manera, el grafo finito es una ventana local de la estructura infinita. Desde el punto de vista geométrico, representa un recorte de la grilla. Desde el punto de vista combinatorio, representa la restricción de la relación de adyacencia global a un subconjunto finito de vértices.
          </p>

          <p>
            Esta observación permite entender mejor la naturaleza del ejemplo inicial. La imagen no describe una red aislada ni un caso excepcional; muestra una pequeña porción de un modelo general cuya formulación natural vive en \( \mathbb{Z}^2 \). Esa continuidad entre el caso finito y el infinito es esencial para que el paso posterior no parezca una ruptura, sino una generalización natural.
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">3.2. Por qué introducir el caso infinito</span>
            <span class="topic-subtitle">Eliminar borde, ganar homogeneidad y estudiar fenómenos globales</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Aunque el grafo finito es útil para visualizar la construcción y estudiar propiedades elementales, resulta insuficiente para comprender el comportamiento global de la red. La razón principal es la presencia de frontera.
          </p>

          <p>
            En el grafo finito aparecen tres tipos distintos de vértices: esquinas, con grado \(2\); bordes no esquina, con grado \(3\); y centro, con grado \(4\). Por tanto, el entorno local de un vértice depende de su posición dentro de la malla. Esta falta de uniformidad altera cualquier proceso dinámico definido sobre el grafo. Un caminante situado en una esquina no tiene las mismas opciones de movimiento que uno situado en el centro. En consecuencia, la estructura finita está fuertemente condicionada por el borde.
          </p>

          <p>
            En muchos problemas matemáticos, sin embargo, interesa estudiar una red homogénea, es decir, una red en la que todos los vértices tengan el mismo entorno local. Esa homogeneidad aparece de manera natural en la grilla infinita \( \mathbb{Z}^2 \), donde todo vértice tiene exactamente cuatro vecinos:
          </p>

          <div class="formula-box">
            <div class="label">Vecinos de un vértice en la red infinita</div>
            \[
            (x+1,y),\qquad (x-1,y),\qquad (x,y+1),\qquad (x,y-1).
            \]
          </div>

          <p>
            No existen esquinas ni bordes. Ningún vértice ocupa una posición privilegiada o marginal. La red se ve igual desde cualquier punto.
          </p>

          <p>
            Introducir el caso infinito se justifica, por tanto, por varias razones matemáticas fundamentales.
          </p>

          <div class="example-box">
            <strong>a) Eliminar efectos de frontera.</strong>
            En el grafo finito, la frontera modifica la estructura local y condiciona la dinámica. En el infinito, esos efectos desaparecen por completo.
          </div>

          <div class="example-box">
            <strong>b) Obtener una red uniforme.</strong>
            La red infinita es regular de grado \(4\). Esta regularidad permite formular resultados más limpios y generales.
          </div>

          <div class="example-box">
            <strong>c) Estudiar comportamientos globales.</strong>
            En una malla pequeña, un caminante está necesariamente confinado. En cambio, en la red infinita puede alejarse sin límite, y por ello adquieren sentido preguntas globales como: ¿regresa al origen?, ¿con qué frecuencia?, ¿cómo decrece la probabilidad de retorno?
          </div>

          <div class="example-box">
            <strong>d) Introducir análisis asintótico.</strong>
            Resultados como
            \[
            p_{2n}(0,0)\sim \frac{1}{\pi n}
            \]
            describen el comportamiento cuando \(n\to\infty\). Tales resultados pertenecen naturalmente al caso infinito, no a una red truncada de tamaño fijo.
          </div>

          <div class="example-box">
            <strong>e) Modelar un medio discreto ideal.</strong>
            La grilla infinita actúa como un modelo canónico del plano discreto. En ella pueden estudiarse difusión, potencial discreto, transporte, percolación y paseos aleatorios sin que intervengan artefactos debidos al borde.
          </div>

          <p>
            Desde un punto de vista conceptual, introducir el caso infinito equivale a pasar de una representación local a una teoría global. El grafo finito sirve para comprender la construcción; el grafo infinito es el objeto sobre el cual se desarrollan los resultados más profundos.
          </p>

          <p>
            Además, la simulación del código que analizaste anteriormente solo adquiere su sentido matemático completo cuando se interpreta sobre la red infinita. La relación con \( \pi \), la probabilidad de retorno al origen y la recurrencia en dos dimensiones son propiedades de la estructura infinita \( \mathbb{Z}^2 \), no de una malla finita \(3\times 3\).
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">3.3. Intuición geométrica de la extensión a todo \( \mathbb{Z}^2 \)</span>
            <span class="topic-subtitle">Quitar el recorte, conservar la vecindad, obtener una red sin frontera</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            La transición del caso finito al caso infinito puede entenderse geométricamente de manera muy natural. En el grafo finito se consideran únicamente los puntos enteros contenidos en el cuadrado
          </p>

          <div class="formula-box">
            <div class="label">Recorte finito del plano discreto</div>
            \[
            0\le x\le 2,
            \qquad
            0\le y\le 2.
            \]
          </div>

          <p>
            Esto produce una malla de tres filas por tres columnas, es decir, una pequeña cuadrícula con nueve vértices. La idea del paso al infinito consiste simplemente en quitar esa restricción sobre las coordenadas y mantener intacta la misma regla de vecindad.
          </p>

          <p>
            En lugar de limitarse a los valores \(0,1,2\), se permite que ambas coordenadas recorran todos los enteros:
          </p>

          <div class="formula-box">
            <div class="label">Paso al caso infinito</div>
            \[
            x\in \mathbb{Z},
            \qquad
            y\in \mathbb{Z}.
            \]
          </div>

          <p>
            Así se obtiene el conjunto de vértices
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de vértices infinito</div>
            \[
            V_\infty=\mathbb{Z}^2=
            \{(x,y):x\in\mathbb{Z},\;y\in\mathbb{Z}\}.
            \]
          </div>

          <p>
            La adyacencia sigue definiéndose del mismo modo:
          </p>

          <div class="formula-box">
            <div class="label">Adyacencia conservada</div>
            \[
            (x,y)\sim(x',y')
            \iff
            (x-x')^2+(y-y')^2=1.
            \]
          </div>

          <p>
            Geométricamente, esto significa que cada punto entero del plano se conecta con sus cuatro vecinos inmediatos: izquierda, derecha, arriba y abajo.
          </p>

          <p>
            Si se quisiera visualizar esta transición, podría pensarse así: la malla \(3\times 3\) de la imagen es solo una pequeña parte visible de una cuadrícula que se prolonga indefinidamente en todas las direcciones. Si se añadieran sucesivamente nuevas filas y columnas alrededor del cuadrado inicial, y este proceso continuara sin detenerse nunca, el resultado sería precisamente la retícula infinita \( \mathbb{Z}^2 \).
          </p>

          <p>
            Esta intuición puede expresarse también en términos de secuencias de recortes crecientes. Para cada \(n\in\mathbb{N}\), consideremos el cuadrado discreto
          </p>

          <div class="formula-box">
            <div class="label">Ventanas crecientes</div>
            \[
            W_n=
            \{(x,y)\in \mathbb{Z}^2:-n\le x\le n,\;-n\le y\le n\}.
            \]
          </div>

          <p>
            Cada \(W_n\) define un subgrafo finito de \( \mathbb{Z}^2 \), y al aumentar \(n\) se obtienen mallas cada vez mayores:
          </p>

          <div class="sequence-box">
            \( W_1 \subset W_2 \subset W_3 \subset \cdots \)
          </div>

          <p>
            Además,
          </p>

          <div class="formula-box">
            <div class="label">Unión de todos los recortes</div>
            \[
            \bigcup_{n\ge 1} W_n=\mathbb{Z}^2.
            \]
          </div>

          <p>
            Esto muestra que el grafo infinito puede entenderse como el límite natural de una familia creciente de grafos finitos. El ejemplo \(3\times 3\) corresponde simplemente a uno de esos recortes.
          </p>

          <p>
            La diferencia esencial está en que, en el caso infinito, la red no tiene borde. Para cualquier vértice \((x,y)\in\mathbb{Z}^2\), los cuatro puntos
          </p>

          <div class="formula-box">
            <div class="label">Vecinos siempre presentes</div>
            \[
            (x+1,y),\qquad (x-1,y),\qquad (x,y+1),\qquad (x,y-1)
            \]
          </div>

          <p>
            pertenecen también a \( \mathbb{Z}^2 \). Por tanto, cada vértice posee siempre cuatro vecinos. No existe una región extrema de la red donde falten conexiones.
          </p>

          <p>
            Esta ausencia de borde tiene una interpretación geométrica y estructural muy profunda: la red infinita es espacialmente homogénea. Si se traslada toda la grilla por un vector \((a,b)\in\mathbb{Z}^2\), la estructura no cambia. En efecto, la aplicación
          </p>

          <div class="formula-box">
            <div class="label">Traslación en la grilla</div>
            \[
            T_{(a,b)}:\mathbb{Z}^2\to\mathbb{Z}^2,
            \qquad
            T_{(a,b)}(x,y)=(x+a,y+b)
            \]
          </div>

          <p>
            preserva la adyacencia, porque
          </p>

          <div class="formula-box">
            <div class="label">Invarianza por traslación</div>
            \[
            ((x+a)-(x'+a))^2+((y+b)-(y'+b))^2
            =
            (x-x')^2+(y-y')^2.
            \]
          </div>

          <p>
            Esto significa que la geometría local es la misma en cualquier parte de la red. Esa propiedad no se cumple en el grafo finito, donde la existencia de borde rompe la simetría traslacional.
          </p>

          <p>
            Por tanto, la intuición geométrica del paso a todo \( \mathbb{Z}^2 \) puede resumirse en tres ideas: se elimina el recorte acotado del plano discreto, se conserva exactamente la misma regla local de conexión y se obtiene una red infinita, uniforme y sin frontera. Esa es la estructura natural sobre la que se formulan los problemas de paseos aleatorios, difusión y retorno al origen en dos dimensiones.
          </p>

        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        El grafo finito es un recorte local; el grafo infinito es la estructura global homogénea.
      </p>
      <p>
        La relación entre ambos grafos no es de oposición, sino de continuidad. El grafo finito de la imagen conserva exactamente la misma regla local de adyacencia que el grafo infinito. La diferencia es que el primero está restringido a una región acotada y, por ello, presenta borde y pérdida de uniformidad.
      </p>
      <p>
        El grafo infinito elimina esa limitación espacial. Al extender la construcción a todo \( \mathbb{Z}^2 \), se obtiene una red sin frontera, regular y homogénea, adecuada para estudiar procesos globales y comportamientos asintóticos. Desde esta perspectiva, el caso finito es una ventana pedagógica sobre una estructura mucho más rica, mientras que el caso infinito es el verdadero objeto matemático de referencia.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_grafo_finito_infinito', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_grafo_finito_infinito') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
# @title  Definición formal del grafo infinito

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Definición formal del grafo infinito</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 920px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 10000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-th"></i>
        Retícula cuadrada infinita
      </div>
      <h1 class="main-title">4. Definición formal del grafo infinito</h1>
      <p class="subtitle">
        La red cuadrada sobre \( \mathbb{Z}^2 \) es el objeto estructural fundamental sobre el cual se definirán el paseo aleatorio, la probabilidad de retorno al origen y la relación asintótica con \( \pi \).
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        El grafo infinito que interesa en este trabajo es la red cuadrada definida sobre el conjunto de puntos enteros del plano. Se trata del objeto estructural fundamental sobre el que se formularán después el paseo aleatorio, la probabilidad de retorno al origen y la relación asintótica con \( \pi \).
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">4.1. Conjunto de vértices</span>
            <span class="topic-subtitle">Todos los puntos enteros del plano como nodos del grafo</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            El conjunto de vértices del grafo infinito es
          </p>

          <div class="formula-box">
            <div class="label">Definición del conjunto de vértices</div>
            \[
            V_\infty=\mathbb{Z}^2=\mathbb{Z}\times\mathbb{Z}.
            \]
          </div>

          <p>
            Esto significa que cada vértice es un par ordenado de números enteros:
          </p>

          <div class="formula-box">
            <div class="label">Forma explícita</div>
            \[
            V_\infty=\{(x,y):x\in\mathbb{Z},\;y\in\mathbb{Z}\}.
            \]
          </div>

          <p>
            Por tanto, cada punto de coordenadas enteras del plano es un vértice del grafo. Ejemplos de vértices son
          </p>

          <div class="formula-box">
            <div class="label">Ejemplos</div>
            \[
            (0,0),\qquad (1,2),\qquad (-3,5),\qquad (7,-4).
            \]
          </div>

          <p>
            A diferencia del caso finito, aquí no hay restricción del tipo \(0\le x\le 2\), \(0\le y\le 2\). El conjunto de vértices se extiende indefinidamente en todas las direcciones del plano.
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">4.2. Conjunto de aristas</span>
            <span class="topic-subtitle">La vecindad inmediata como regla global de conexión</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            El conjunto de aristas viene dado por
          </p>

          <div class="formula-box">
            <div class="label">Definición del conjunto de aristas</div>
            \[
            E_\infty=
            \Bigl\{
            \{(x,y),(x',y')\}\subseteq \mathbb{Z}^2:
            (x-x')^2+(y-y')^2=1
            \Bigr\}.
            \]
          </div>

          <p>
            Esta definición indica que dos vértices están unidos por una arista si la distancia euclidiana entre ellos es exactamente \(1\). Como las coordenadas son enteras, eso equivale a decir que son vecinos inmediatos horizontales o verticales.
          </p>

          <p>
            De forma equivalente,
          </p>

          <div class="formula-box">
            <div class="label">Formulación equivalente</div>
            \[
            (x,y)\sim(x',y')
            \iff
            (x',y')\in\{(x+1,y),(x-1,y),(x,y+1),(x,y-1)\}.
            \]
          </div>

          <p>
            Por ejemplo,
          </p>

          <div class="example-box">
            <strong>Ejemplo.</strong>
            \[
            (0,0)\sim(1,0),\qquad
            (0,0)\sim(-1,0),\qquad
            (0,0)\sim(0,1),\qquad
            (0,0)\sim(0,-1),
            \]
            pero
            \[
            (0,0)\not\sim(1,1),
            \]
            porque
            \[
            (0-1)^2+(0-1)^2=2\neq 1.
            \]
          </div>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">4.3. Definición del grafo</span>
            <span class="topic-subtitle">La formulación completa del objeto matemático</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Con estas definiciones, el grafo infinito queda dado por
          </p>

          <div class="formula-box">
            <div class="label">Definición formal</div>
            \[
            G_\infty=(V_\infty,E_\infty).
            \]
          </div>

          <p>
            Es decir,
          </p>

          <div class="formula-box">
            <div class="label">Expresión completa</div>
            \[
            G_\infty=
            \left(
            \mathbb{Z}^2,
            \Bigl\{
            \{(x,y),(x',y')\}\subseteq \mathbb{Z}^2:
            (x-x')^2+(y-y')^2=1
            \Bigr\}
            \right).
            \]
          </div>

          <p>
            Este es el grafo de la red cuadrada infinita del plano discreto.
          </p>

        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">4.4. Formulación equivalente mediante vecinos inmediatos</span>
            <span class="topic-subtitle">La versión más útil para procesos dinámicos y paseos aleatorios</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">

          <p>
            Para cada vértice \((x,y)\in\mathbb{Z}^2\), se define su conjunto de vecinos por
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de vecinos</div>
            \[
            N(x,y)=\{(x+1,y),(x-1,y),(x,y+1),(x,y-1)\}.
            \]
          </div>

          <p>
            Entonces la relación de adyacencia puede expresarse como
          </p>

          <div class="formula-box">
            <div class="label">Equivalencia con la adyacencia</div>
            \[
            (x',y')\in N(x,y)\iff (x,y)\sim(x',y').
            \]
          </div>

          <p>
            Esta formulación es equivalente a la definición mediante la distancia \(1\) y resulta especialmente útil porque hace explícitos los cuatro movimientos elementales posibles, conecta directamente con el paseo aleatorio, muestra de inmediato que cada vértice tiene grado \(4\) y resalta la homogeneidad del grafo.
          </p>

          <p>
            En particular,
          </p>

          <div class="formula-box">
            <div class="label">Grado de cada vértice</div>
            \[
            |N(x,y)|=4
            \qquad
            \text{para todo } (x,y)\in\mathbb{Z}^2.
            \]
          </div>

        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        La red cuadrada infinita como grafo sobre \( \mathbb{Z}^2 \).
      </p>
      <p>
        La definición formal del grafo infinito consiste en tomar todos los puntos enteros del plano como vértices y conectar únicamente aquellos que son vecinos inmediatos. Esta construcción produce una grilla regular infinita, sin borde, en la que cada vértice tiene exactamente cuatro vecinos.
      </p>
      <p>
        La importancia de esta formulación es doble. Por un lado, aporta rigor: el objeto queda completamente determinado por \(V_\infty\) y \(E_\infty\). Por otro lado, prepara el terreno para definir procesos dinámicos sobre el grafo, especialmente el paseo aleatorio simple. Cuando un caminante se mueve desde \((x,y)\) a uno de sus vecinos inmediatos con probabilidad uniforme, está recorriendo precisamente este grafo.
      </p>
      <p>
        Por ello, \(G_\infty\) no es solo una red geométrica: es el soporte matemático de toda la teoría posterior.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_def_formal_grafo_infinito', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_def_formal_grafo_infinito') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
# @title Sección 5 con acordeones: Del grafo finito al grafo infinito en Z^2 (Ejecutar esta celda)

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Del grafo finito al grafo infinito en Z²</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
      --table-header: rgba(14, 165, 233, 0.12);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
      --table-header: rgba(255, 255, 255, 0.08);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 920px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 14000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    table {
      width: 100%;
      border-collapse: collapse;
      margin: 1rem 0 1.1rem;
      overflow: hidden;
      border-radius: 14px;
      box-shadow: var(--shadow-soft);
    }

    th, td {
      border: 1px solid var(--border-color);
      padding: 0.85rem 0.9rem;
      text-align: left;
      color: var(--text-secondary);
      vertical-align: top;
    }

    th {
      background: var(--table-header);
      color: var(--text-primary);
      font-weight: 700;
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-expand"></i>
        Generalización al plano discreto completo
      </div>
      <h1 class="main-title">5. Del grafo finito al grafo infinito en \( \mathbb{Z}^2 \)</h1>
      <p class="subtitle">
        El paso del grafo finito al grafo infinito constituye una generalización natural desde una estructura discreta acotada hacia un modelo homogéneo definido sobre todo el plano entero.
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        Mientras que en el caso finito los vértices se restringen a una región determinada de la grilla, en el caso infinito desaparecen esas limitaciones espaciales y la construcción se extiende a la totalidad de \( \mathbb{Z}^2 \). Esta transición no solo implica un aumento en la cantidad de vértices y aristas, sino también un cambio conceptual profundo: se pasa de una red con borde y efectos de contorno a una estructura uniforme, sin fronteras, adecuada para el estudio de propiedades globales y fenómenos asintóticos.
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.1. De un subconjunto acotado a \( \mathbb{Z}^2 \)</span>
            <span class="topic-subtitle">Del recorte finito del plano a la red entera completa</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            En el caso finito, el conjunto de vértices se obtiene restringiendo las coordenadas a un intervalo acotado. Por ejemplo, en la grilla \(3\times 3\) se considera el conjunto
          </p>

          <div class="formula-box">
            <div class="label">Conjunto finito de vértices</div>
            \[
            V=\{(x,y)\in\mathbb{Z}^2:0\le x\le 2,\;0\le y\le 2\}.
            \]
          </div>

          <p>
            Esta definición describe una porción discreta y finita del plano entero. En cambio, en el caso infinito tales restricciones desaparecen completamente y el conjunto de vértices pasa a ser
          </p>

          <div class="formula-box">
            <div class="label">Conjunto infinito de vértices</div>
            \[
            V_\infty=\mathbb{Z}^2.
            \]
          </div>

          <p>
            De este modo, cada par ordenado de enteros representa un vértice del grafo. La estructura obtenida ya no corresponde a una ventana local del espacio discreto, sino al reticulado bidimensional completo. Esta extensión conserva la idea de vecindad local del caso finito, pero la sitúa en un contexto mucho más amplio, donde no existe una región privilegiada ni una frontera exterior que delimite el sistema.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.2. Eliminación de restricciones de borde</span>
            <span class="topic-subtitle">La desaparición de la frontera y la homogeneidad local</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            La principal diferencia geométrica entre ambos casos radica en la presencia o ausencia de borde. En la grilla finita, las condiciones
          </p>

          <div class="formula-box">
            <div class="label">Restricciones de borde</div>
            \[
            0\le x\le 2,\qquad 0\le y\le 2
            \]
          </div>

          <p>
            imponen una frontera que modifica la estructura local de los vértices cercanos al contorno. Como consecuencia, los vértices de esquina tienen menos vecinos que los vértices del centro, y los vértices ubicados en los bordes también presentan una conectividad reducida.
          </p>

          <p>
            En el grafo infinito, esas restricciones desaparecen por completo. Ya no existe frontera geométrica, por lo que todo vértice se encuentra en una situación local idéntica. La relación de adyacencia se define de la misma manera que en el caso finito, es decir,
          </p>

          <div class="formula-box">
            <div class="label">Relación de adyacencia global</div>
            \[
            E_\infty=
            \Bigl\{
            \{(x,y),(x',y')\}\subseteq\mathbb{Z}^2:
            (x-x')^2+(y-y')^2=1
            \Bigr\}.
            \]
          </div>

          <p>
            Esta condición expresa que dos vértices están unidos por una arista si y solo si son vecinos inmediatos en dirección horizontal o vertical. En consecuencia, cada vértice \((x,y)\) tiene exactamente cuatro vecinos:
          </p>

          <div class="formula-box">
            <div class="label">Vecinos inmediatos</div>
            \[
            (x+1,y),\qquad (x-1,y),\qquad (x,y+1),\qquad (x,y-1).
            \]
          </div>

          <p>
            La eliminación del borde produce una red completamente homogénea desde el punto de vista local, propiedad que resulta fundamental en el análisis teórico del grafo infinito.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.3. Cambio de cardinalidad</span>
            <span class="topic-subtitle">Del conteo finito al régimen infinito numerable</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            En el caso finito, tanto el conjunto de vértices como el conjunto de aristas tienen cardinalidad finita. Por ejemplo, la cuadrícula \(3\times 3\) posee exactamente 9 vértices y un número finito de conexiones entre ellos. En el caso infinito, en cambio, el conjunto de vértices deja de ser finito y pasa a ser infinito numerable, ya que \( \mathbb{Z}^2 \) puede ponerse en correspondencia biyectiva con \( \mathbb{N} \).
          </p>

          <p>
            Este cambio de cardinalidad modifica el tipo de herramientas matemáticas apropiadas para su estudio. En grafos finitos es habitual trabajar con conteos exactos, matrices de adyacencia finitas, grados, caminos mínimos y propiedades combinatorias directas. En el contexto infinito, por el contrario, aparecen preguntas relacionadas con el comportamiento global de la red, el crecimiento, las distancias a gran escala, la teoría espectral, la recurrencia de trayectorias y los fenómenos probabilísticos asociados a procesos definidos sobre el grafo.
          </p>

          <p>
            Por lo tanto, el paso al infinito no consiste únicamente en agregar más vértices, sino en cambiar de régimen matemático: se abandona un objeto finitamente controlable para ingresar en un marco donde adquieren relevancia el análisis asintótico y las propiedades estructurales globales.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.4. Notación de subgrafo inducido</span>
            <span class="topic-subtitle">La grilla finita como ventana observacional dentro de la red infinita</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Una forma precisa de relacionar ambos modelos consiste en interpretar el grafo finito como un subgrafo inducido del grafo infinito. Si se define
          </p>

          <div class="formula-box">
            <div class="label">Subconjunto finito</div>
            \[
            W=
            \{(x,y)\in\mathbb{Z}^2:0\le x\le 2,\;0\le y\le 2\},
            \]
          </div>

          <p>
            entonces la grilla finita \(3\times 3\) puede describirse como
          </p>

          <div class="formula-box">
            <div class="label">Subgrafo inducido</div>
            \[
            G_{3\times 3}=G_\infty[W],
            \]
          </div>

          <p>
            donde \(G_\infty[W]\) representa el subgrafo inducido por el subconjunto \(W\) dentro del grafo infinito \(G_\infty\). Esta notación significa que se toman como vértices exactamente los elementos de \(W\), y como aristas todas aquellas conexiones del grafo infinito cuyos dos extremos pertenecen a ese conjunto.
          </p>

          <p>
            Esta formulación es conceptualmente muy importante porque permite entender al grafo finito no como un objeto aislado, sino como una restricción local de una estructura más amplia. En otras palabras, una grilla finita puede verse como una ventana observacional dentro de la red infinita. Esta idea resulta especialmente útil cuando se pretende aproximar fenómenos definidos en medios infinitos mediante simulaciones o análisis en dominios finitos, ya que proporciona una relación formal entre ambos contextos.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.5. Notación del paseo aleatorio y de las probabilidades de retorno</span>
            <span class="topic-subtitle">La conexión entre el grafo infinito y los procesos estocásticos</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El estudio del grafo infinito sobre \( \mathbb{Z}^2 \) adquiere una gran riqueza cuando se consideran procesos dinámicos o aleatorios sobre sus vértices. El ejemplo más importante es el paseo aleatorio simple. En este contexto, se introduce una sucesión de variables aleatorias
          </p>

          <div class="formula-box">
            <div class="label">Proceso aleatorio</div>
            \[
            (X_n)_{n\ge 0},
            \]
          </div>

          <p>
            donde \(X_n\) representa la posición del caminante en el instante \(n\). Si el proceso comienza en un vértice inicial \(x\), se escribe \(X_0=x\). En particular, cuando el recorrido se inicia en el origen, se toma
          </p>

          <div class="formula-box">
            <div class="label">Condición inicial en el origen</div>
            \[
            X_0=(0,0).
            \]
          </div>

          <p>
            La dinámica del paseo aleatorio simple consiste en que, en cada paso, el caminante se desplaza con igual probabilidad hacia uno de los cuatro vecinos inmediatos del vértice actual. Esto significa que, desde cualquier vértice, la probabilidad de moverse en cada una de las cuatro direcciones posibles es \(1/4\). La geometría uniforme del grafo infinito hace que esta definición sea natural y estable en todo el espacio.
          </p>

          <p>
            Para describir matemáticamente la evolución del proceso se utiliza la notación de probabilidades de transición
          </p>

          <div class="formula-box">
            <div class="label">Probabilidades de transición</div>
            \[
            p_n(x,y)=\mathbb{P}(X_n=y\mid X_0=x),
            \]
          </div>

          <p>
            que representa la probabilidad de que el paseo se encuentre en el vértice \(y\) luego de \(n\) pasos, habiendo comenzado en \(x\). En particular, la probabilidad de retorno al origen en el tiempo \(n\) se expresa como
          </p>

          <div class="formula-box">
            <div class="label">Probabilidad de retorno al origen</div>
            \[
            p_n(0,0)=\mathbb{P}(X_n=(0,0)\mid X_0=(0,0)).
            \]
          </div>

          <p>
            Esta notación es esencial en el estudio del grafo infinito porque permite analizar propiedades probabilísticas profundas de la red. En el caso de \( \mathbb{Z}^2 \), el paseo aleatorio simple es recurrente, lo que significa que retorna al origen con probabilidad igual a 1. Además, las probabilidades de retorno presentan leyes asintóticas características que reflejan la naturaleza del espacio bidimensional discreto. Por ello, la notación \(X_n\), \(p_n(x,y)\) y en particular \(p_n(0,0)\), adquiere pleno sentido en el caso infinito y constituye una herramienta central para conectar la teoría de grafos con la probabilidad y el análisis.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.6. Diferencias estructurales entre el caso finito y el caso infinito</span>
            <span class="topic-subtitle">Comparación de borde, grados, técnicas y propiedades globales</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            La comparación entre ambos modelos permite destacar varias diferencias estructurales fundamentales. En primer lugar, el grafo finito posee borde, mientras que el grafo infinito carece completamente de frontera. En segundo lugar, en la grilla finita los grados de los vértices no son uniformes: las esquinas tienen grado 2, los vértices de borde grado 3 y el vértice central grado 4. En contraste, en el grafo infinito todos los vértices tienen grado exactamente 4.
          </p>

          <p>
            En tercer lugar, el análisis del caso finito se apoya principalmente en herramientas combinatorias, geométricas y matriciales de dimensión finita, mientras que el caso infinito exige técnicas adicionales vinculadas con el análisis asintótico, la teoría espectral y la probabilidad. Finalmente, aunque el grafo finito puede entenderse como una aproximación local del infinito, no reproduce completamente sus propiedades globales. Fenómenos como la recurrencia del paseo aleatorio, la ausencia de borde y la homogeneidad total del medio solo se manifiestan con plenitud en el contexto infinito.
          </p>

          <table>
            <thead>
              <tr>
                <th>Aspecto</th>
                <th>Caso finito</th>
                <th>Caso infinito</th>
              </tr>
            </thead>
            <tbody>
              <tr>
                <td>Borde</td>
                <td>Presente</td>
                <td>Ausente</td>
              </tr>
              <tr>
                <td>Grados</td>
                <td>No uniformes</td>
                <td>Todos iguales a 4</td>
              </tr>
              <tr>
                <td>Cardinalidad</td>
                <td>Finita</td>
                <td>Infinita numerable</td>
              </tr>
              <tr>
                <td>Técnicas dominantes</td>
                <td>Combinatoria y matrices finitas</td>
                <td>Análisis asintótico, espectral y probabilidad</td>
              </tr>
              <tr>
                <td>Fenómenos globales</td>
                <td>Parcialmente visibles</td>
                <td>Se manifiestan plenamente</td>
              </tr>
            </tbody>
          </table>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">5.7. Interpretación conceptual del pasaje al infinito</span>
            <span class="topic-subtitle">La conservación de la regla local y la desaparición de toda frontera</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Desde un punto de vista conceptual, pasar del grafo finito al grafo infinito significa conservar la misma regla local de vecindad, pero eliminar toda restricción geométrica externa. El resultado es una red discreta bidimensional uniforme, sin borde, infinita y regular, que sirve como modelo canónico del espacio discreto plano.
          </p>

          <p>
            Esta transición es fundamental porque permite comprender que muchas propiedades observadas en grafos finitos son solo manifestaciones locales de una estructura más general. El caso finito ofrece una representación accesible, visualizable y computable, mientras que el caso infinito proporciona el marco natural para estudiar propiedades globales, trayectorias largas, comportamiento de retorno, difusión y simetrías espaciales.
          </p>

          <p>
            En síntesis, el grafo infinito sobre \( \mathbb{Z}^2 \) puede interpretarse como la prolongación natural de la malla finita hacia todo el plano entero. El primero hereda del segundo la noción elemental de vecindad ortogonal, pero la ausencia de frontera y la infinitud de la red lo convierten en un objeto matemático de mayor profundidad teórica, especialmente relevante en teoría de grafos, procesos estocásticos y modelos discretos del espacio.
          </p>
        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        El paso al infinito no cambia la regla local de conexión, pero transforma profundamente la naturaleza del objeto matemático.
      </p>
      <p>
        En la grilla finita, la vecindad ortogonal ya está presente, pero aparece afectada por el borde y por la desigualdad de grados entre esquinas, bordes y centro. En la red infinita, esas limitaciones desaparecen: todos los vértices tienen la misma estructura local y el medio se vuelve homogéneo.
      </p>
      <p>
        Esta diferencia convierte al grafo infinito en el marco natural para estudiar fenómenos globales, paseos aleatorios, recurrencia, probabilidades de retorno y comportamientos asintóticos. El grafo finito sigue siendo valioso como modelo local y pedagógico, pero el grafo infinito es el objeto estructural de referencia.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_del_finito_al_infinito_z2', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_del_finito_al_infinito_z2') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
# @title Propiedades del grafo infinito

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Propiedades del grafo infinito</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
      --table-header: rgba(14, 165, 233, 0.12);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
      --table-header: rgba(255, 255, 255, 0.08);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 940px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 16000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    table {
      width: 100%;
      border-collapse: collapse;
      margin: 1rem 0 1.1rem;
      overflow: hidden;
      border-radius: 14px;
      box-shadow: var(--shadow-soft);
    }

    th, td {
      border: 1px solid var(--border-color);
      padding: 0.85rem 0.9rem;
      text-align: left;
      color: var(--text-secondary);
      vertical-align: top;
    }

    th {
      background: var(--table-header);
      color: var(--text-primary);
      font-weight: 700;
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-network-wired"></i>
        Estructura global del plano discreto
      </div>
      <h1 class="main-title">6. Propiedades del grafo infinito</h1>
      <p class="subtitle">
        El grafo infinito sobre \( \mathbb{Z}^2 \) constituye uno de los ejemplos más importantes de la teoría de grafos discretos y de los procesos aleatorios sobre redes, combinando simplicidad de definición con una estructura regular, simétrica y profundamente rica.
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        A diferencia del caso finito, donde aparecen efectos de borde y restricciones geométricas, el grafo infinito presenta uniformidad local en todos sus vértices y permite analizar fenómenos globales, asintóticos y probabilísticos con gran profundidad. En esta sección se describen sus propiedades estructurales más relevantes.
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.1. Infinitud numerable</span>
            <span class="topic-subtitle">Un grafo infinito, pero todavía tratable en términos discretos</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El conjunto de vértices del grafo infinito es
          </p>

          <div class="formula-box">
            <div class="label">Conjunto de vértices</div>
            \[
            V_\infty=\mathbb{Z}^2.
            \]
          </div>

          <p>
            Dado que \( \mathbb{Z} \) es un conjunto infinito numerable, también lo es \( \mathbb{Z}^2 \). En consecuencia, aunque el grafo posee infinitos vértices, estos pueden enumerarse mediante una sucesión, lo cual permite tratarlos con herramientas rigurosas de combinatoria, teoría de grafos y probabilidad.
          </p>

          <p>
            La numerabilidad resulta especialmente importante desde el punto de vista formal, ya que garantiza que el conjunto de estados de procesos definidos sobre el grafo, como el paseo aleatorio, siga siendo accesible dentro del marco clásico de espacios discretos numerables. Así, la infinitud del grafo no impide un tratamiento matemático preciso, sino que abre la puerta al estudio de propiedades globales que no aparecen en estructuras finitas.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.2. Conectividad</span>
            <span class="topic-subtitle">Todo par de vértices puede unirse mediante un camino finito</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El grafo infinito es conexo. Esto significa que para cualquier par de vértices \(u=(x_1,y_1)\) y \(v=(x_2,y_2)\) existe siempre un camino finito que une \(u\) con \(v\).
          </p>

          <p>
            La razón es inmediata a partir de la geometría de la red: desde \((x_1,y_1)\) puede avanzarse horizontalmente hasta alcanzar la coordenada \(x_2\), y luego verticalmente hasta llegar a \(y_2\), o viceversa. En términos más explícitos, es posible construir un camino realizando \(|x_2-x_1|\) pasos horizontales y \(|y_2-y_1|\) pasos verticales. Por lo tanto, todo par de vértices está conectado por un camino de longitud finita.
          </p>

          <p>
            Esta propiedad muestra que el grafo constituye una única componente conexa. Además, la distancia natural entre dos vértices coincide con la distancia Manhattan:
          </p>

          <div class="formula-box">
            <div class="label">Distancia geodésica</div>
            \[
            d\big((x_1,y_1),(x_2,y_2)\big)=|x_2-x_1|+|y_2-y_1|.
            \]
          </div>

          <p>
            Por ello, la conectividad del grafo no solo asegura la existencia de caminos, sino que además permite describir de forma explícita su longitud mínima.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.3. Regularidad de grado 4</span>
            <span class="topic-subtitle">Cada vértice tiene exactamente cuatro vecinos</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Cada vértice \((x,y)\in\mathbb{Z}^2\) tiene exactamente cuatro vecinos inmediatos:
          </p>

          <div class="formula-box">
            <div class="label">Vecinos inmediatos</div>
            \[
            (x+1,y),\qquad (x-1,y),\qquad (x,y+1),\qquad (x,y-1).
            \]
          </div>

          <p>
            De aquí se sigue que el grafo infinito es regular de grado 4. Es decir, todos los vértices tienen exactamente el mismo número de vecinos.
          </p>

          <p>
            La regularidad es una propiedad fundamental porque expresa la uniformidad combinatoria de la red. Ningún vértice posee una posición privilegiada desde el punto de vista local. Esta homogeneidad simplifica notablemente el estudio estructural y también el análisis de procesos dinámicos definidos sobre el grafo, ya que la regla de evolución es idéntica en todos los puntos.
          </p>

          <p>
            En particular, para el paseo aleatorio simple, la regularidad implica que desde cualquier vértice la probabilidad de transición a cada uno de sus vecinos es la misma, a saber, \(1/4\). Esta simetría local es una de las razones por las cuales \( \mathbb{Z}^2 \) constituye un modelo central en probabilidad discreta.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.4. Bipartición</span>
            <span class="topic-subtitle">La paridad de \(x+y\) separa naturalmente los vértices en dos clases</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El grafo infinito es bipartito. Esto significa que su conjunto de vértices puede dividirse en dos clases disjuntas tales que toda arista une un vértice de una clase con uno de la otra.
          </p>

          <p>
            Una partición natural se obtiene considerando la paridad de la suma de coordenadas. Definimos
          </p>

          <div class="formula-box">
            <div class="label">Clases de bipartición</div>
            \[
            V_0=\{(x,y)\in\mathbb{Z}^2:x+y \text{ es par}\},
            \]
            \[
            V_1=\{(x,y)\in\mathbb{Z}^2:x+y \text{ es impar}\}.
            \]
          </div>

          <p>
            Entonces,
          </p>

          <div class="formula-box">
            <div class="label">Descomposición disjunta</div>
            \[
            \mathbb{Z}^2=V_0\cup V_1,\qquad V_0\cap V_1=\varnothing.
            \]
          </div>

          <p>
            Si dos vértices son adyacentes, difieren en una unidad en una sola coordenada, por lo que la suma \(x+y\) cambia necesariamente de paridad. En consecuencia, cada arista conecta siempre un vértice de \(V_0\) con uno de \(V_1\), y nunca dos vértices de la misma clase.
          </p>

          <p>
            Esta bipartición tiene múltiples consecuencias. Por ejemplo, implica que todo ciclo del grafo tiene longitud par y explica por qué un paseo aleatorio que parte del origen solo puede regresar a él en tiempos pares. Se trata, por tanto, de una propiedad estructural con repercusiones directas en el análisis combinatorio y probabilístico.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.5. Invarianza por traslación</span>
            <span class="topic-subtitle">La red tiene la misma estructura desde cualquier vértice</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Una propiedad central del grafo infinito es su invarianza por traslación. Si se considera un vector \((a,b)\in\mathbb{Z}^2\) y se traslada cada vértice \((x,y)\) al vértice \((x+a,y+b)\), la estructura del grafo permanece inalterada.
          </p>

          <p>
            Formalmente, la aplicación
          </p>

          <div class="formula-box">
            <div class="label">Traslación</div>
            \[
            T_{(a,b)}(x,y)=(x+a,y+b)
            \]
          </div>

          <p>
            es un automorfismo del grafo. En efecto, si dos vértices \((x,y)\) y \((x',y')\) son adyacentes, entonces
          </p>

          <div class="formula-box">
            <div class="label">Preservación de la adyacencia</div>
            \[
            (x-x')^2+(y-y')^2=1,
            \]
            y por consiguiente también se cumple
            \[
            ((x+a)-(x'+a))^2+((y+b)-(y'+b))^2=1.
            \]
          </div>

          <p>
            Esto demuestra que la relación de adyacencia se conserva bajo cualquier traslación entera del plano.
          </p>

          <p>
            La importancia de esta propiedad radica en que el grafo tiene la misma apariencia desde cualquier vértice. No existe un centro natural ni una región distinguida. Esta homogeneidad espacial resulta esencial en el estudio de operadores sobre el grafo, del paseo aleatorio y de argumentos basados en simetría.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.6. Simetrías geométricas</span>
            <span class="topic-subtitle">Reflexiones y rotaciones compatibles con la estructura cuadrada</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Además de las traslaciones, el grafo infinito posee otras simetrías geométricas vinculadas con la estructura cuadrada de la malla. En particular, es invariante bajo reflexiones respecto de ejes coordenados, reflexiones respecto de diagonales y rotaciones de ángulo \(90^\circ\), \(180^\circ\) y \(270^\circ\) alrededor de un vértice.
          </p>

          <p>
            Estas transformaciones preservan la red porque mantienen las distancias de vecindad definidas por movimientos horizontales y verticales unitarios. Por ejemplo, una rotación de \(90^\circ\) transforma el vértice \((x,y)\) en \((-y,x)\), y esta transformación conserva la adyacencia. Del mismo modo, reflexiones como \((x,y)\mapsto(-x,y)\) o \((x,y)\mapsto(y,x)\) no alteran la estructura del grafo.
          </p>

          <p>
            Estas simetrías no son meramente ornamentales. Cumplen un papel decisivo en numerosos argumentos combinatorios y probabilísticos, ya que permiten reducir casos, comparar trayectorias y deducir igualdades de probabilidad entre configuraciones geométricamente equivalentes. La gran cantidad de automorfismos del grafo refleja su elevado grado de regularidad y uniformidad.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.7. Ausencia de borde</span>
            <span class="topic-subtitle">Todos los vértices comparten el mismo entorno local</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            A diferencia del caso finito, el grafo infinito no tiene borde. Ningún vértice ocupa una posición extrema o especial dentro de la red. Todos los vértices poseen exactamente cuatro vecinos y comparten el mismo entorno local.
          </p>

          <p>
            Esta ausencia de frontera tiene consecuencias estructurales profundas. En primer lugar, elimina las irregularidades locales que en el caso finito distinguían a esquinas, bordes y vértices interiores. En segundo lugar, hace que las propiedades del grafo sean verdaderamente homogéneas en toda la red. En tercer lugar, simplifica el análisis de procesos dinámicos, puesto que no existen efectos de contorno que alteren la evolución del sistema en ciertas zonas.
          </p>

          <p>
            Desde el punto de vista conceptual, la ausencia de borde convierte al grafo infinito en un modelo idealizado del espacio discreto bidimensional: un medio uniforme, ilimitado y localmente idéntico en todas partes.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.8. Diferencias estructurales frente al caso finito</span>
            <span class="topic-subtitle">Comparación entre la malla acotada y la red global homogénea</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            La comparación entre el grafo infinito y su versión finita permite destacar varias diferencias estructurales esenciales. En el caso finito, la presencia de borde genera vértices con grados distintos según su posición: las esquinas tienen grado 2, los vértices laterales grado 3 y solo algunos vértices interiores alcanzan el grado máximo permitido por la vecindad ortogonal. En cambio, en el caso infinito todos los vértices tienen grado 4, sin excepción.
          </p>

          <p>
            Asimismo, en el grafo finito muchas propiedades dependen del tamaño de la ventana observada. La forma del dominio, la ubicación de los bordes y la dimensión de la grilla influyen directamente en caminos, distancias, conteos y trayectorias. En el caso infinito, por el contrario, esas dependencias desaparecen y emergen fenómenos globales y asintóticos que no pueden captarse completamente en una red acotada.
          </p>

          <p>
            Otra diferencia importante radica en el tipo de problemas que cada modelo permite estudiar. El caso finito es especialmente adecuado para representaciones concretas, cálculos exactos, matrices de tamaño finito y visualización geométrica. El caso infinito, en cambio, es el marco natural para analizar recurrencia, difusión, comportamiento a largo plazo, crecimiento de trayectorias y leyes límite de procesos aleatorios.
          </p>

          <p>
            En este sentido, el grafo finito puede interpretarse como una aproximación local o truncada del infinito, útil para ilustrar la estructura básica de vecindad. Sin embargo, la riqueza matemática plena del modelo aparece cuando se considera la red completa sobre \( \mathbb{Z}^2 \), donde la homogeneidad, la ausencia de borde y la infinitud permiten desarrollar una teoría más profunda y general.
          </p>

          <table>
            <thead>
              <tr>
                <th>Aspecto</th>
                <th>Grafo finito</th>
                <th>Grafo infinito</th>
              </tr>
            </thead>
            <tbody>
              <tr>
                <td>Borde</td>
                <td>Presente</td>
                <td>Ausente</td>
              </tr>
              <tr>
                <td>Grados</td>
                <td>Variables según la posición</td>
                <td>Todos iguales a 4</td>
              </tr>
              <tr>
                <td>Cardinalidad</td>
                <td>Finita</td>
                <td>Infinita numerable</td>
              </tr>
              <tr>
                <td>Dependencia del tamaño</td>
                <td>Alta</td>
                <td>No depende de una ventana finita</td>
              </tr>
              <tr>
                <td>Problemas naturales</td>
                <td>Conteos y visualización local</td>
                <td>Recurrencia, difusión, análisis asintótico</td>
              </tr>
            </tbody>
          </table>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">6.9. Síntesis estructural</span>
            <span class="topic-subtitle">Una red discreta regular, homogénea y central para la teoría</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            En conjunto, las propiedades anteriores muestran que el grafo infinito sobre \( \mathbb{Z}^2 \) es una estructura discreta altamente regular, conexa, bipartita, numerable, sin borde y compatible con numerosas simetrías geométricas. Estas características lo convierten en un objeto central tanto en teoría de grafos como en probabilidad, análisis discreto y física matemática.
          </p>

          <p>
            Su importancia no reside únicamente en la simplicidad de su definición, sino en el hecho de que constituye un modelo canónico del espacio discreto bidimensional. Gracias a su uniformidad local y global, este grafo sirve como escenario natural para estudiar problemas de conectividad, distancias, recorridos, operadores discretos, paseos aleatorios y fenómenos de retorno. Por ello, comprender sus propiedades estructurales básicas es un paso esencial para cualquier análisis posterior de tipo combinatorio, geométrico o estocástico.
          </p>
        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        El grafo infinito sobre \( \mathbb{Z}^2 \) es una red discreta uniforme, sin borde, regular de grado 4 y compatible con una gran cantidad de simetrías.
      </p>
      <p>
        Su numerabilidad permite tratarlo rigurosamente como espacio discreto, su conectividad asegura caminos entre cualesquiera dos vértices, su bipartición organiza la red según la paridad de las coordenadas y su invarianza por traslación muestra que no existe un centro privilegiado.
      </p>
      <p>
        Estas propiedades hacen de \( \mathbb{Z}^2 \) el soporte natural para el estudio de paseos aleatorios, difusión, recurrencia y fenómenos asintóticos. El paso desde el caso finito hacia esta estructura global es precisamente el que abre la puerta a los resultados más profundos del modelo.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_propiedades_grafo_infinito', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_propiedades_grafo_infinito') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
# @title El paseo aleatorio sobre Z^2

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>El paseo aleatorio sobre Z²</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
      --table-header: rgba(14, 165, 233, 0.12);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
      --table-header: rgba(255, 255, 255, 0.08);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 940px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 18000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-random"></i>
        Proceso estocástico sobre la red cuadrada
      </div>
      <h1 class="main-title">7. El paseo aleatorio sobre \( \mathbb{Z}^2 \)</h1>
      <p class="subtitle">
        El paseo aleatorio simple sobre \( \mathbb{Z}^2 \) constituye uno de los modelos más importantes de la probabilidad discreta y de los procesos estocásticos sobre grafos.
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        Su interés radica en que describe, de manera elemental pero profunda, el movimiento de una partícula que se desplaza al azar sobre la red bidimensional infinita, eligiendo en cada instante uno de sus vecinos inmediatos con igual probabilidad. A pesar de la sencillez de su definición, este proceso exhibe propiedades matemáticas notables, entre ellas la estructura de cadena de Markov, la restricción de paridad en los retornos y la recurrencia característica de la dimensión dos.
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.1. Definición formal del proceso</span>
            <span class="topic-subtitle">La trayectoria aleatoria como sucesión de posiciones en la red</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Un paseo aleatorio simple sobre \( \mathbb{Z}^2 \) es un proceso estocástico
          </p>

          <div class="formula-box">
            <div class="label">Proceso estocástico</div>
            \[
            (X_n)_{n\ge 0},
            \]
          </div>

          <p>
            definido sobre el conjunto de estados \( \mathbb{Z}^2 \), donde \(X_n\) representa la posición del caminante en el instante \(n\). Se parte de una posición inicial \(X_0=x_0\), con \(x_0\in\mathbb{Z}^2\), y en cada paso el proceso se desplaza a uno de los cuatro vecinos inmediatos del vértice actual.
          </p>

          <p>
            Cuando se estudian propiedades de retorno, es habitual tomar como punto de partida el origen, es decir,
          </p>

          <div class="formula-box">
            <div class="label">Condición inicial</div>
            \[
            X_0=(0,0).
            \]
          </div>

          <p>
            El proceso queda así completamente determinado por la posición inicial y por la regla probabilística de transición entre vértices adyacentes. En cada tiempo discreto \(n\), el caminante ocupa exactamente un vértice de la red, y la sucesión de estas posiciones describe una trayectoria aleatoria sobre el grafo infinito.
          </p>

          <p>
            Desde un punto de vista intuitivo, el paseo aleatorio modela una evolución sin dirección preferencial, en la que cada movimiento local depende únicamente de la estructura de vecindad de la malla cuadrada. Esta simplicidad convierte al proceso en un modelo canónico para estudiar difusión discreta, exploración aleatoria del espacio y retorno a estados previamente visitados.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.2. Regla de transición</span>
            <span class="topic-subtitle">Cuatro movimientos posibles, todos equiprobables</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            La regla de transición del paseo aleatorio simple se basa directamente en la geometría local del grafo infinito. Si en el instante \(n\) el caminante se encuentra en el vértice \((x,y)\), entonces en el instante siguiente pasa a uno de los cuatro vértices
          </p>

          <div class="formula-box">
            <div class="label">Movimientos elementales</div>
            \[
            (x+1,y),\qquad (x-1,y),\qquad (x,y+1),\qquad (x,y-1),
            \]
          </div>

          <p>
            cada uno con probabilidad
          </p>

          <div class="formula-box">
            <div class="label">Probabilidad de cada movimiento</div>
            \[
            \frac{1}{4}.
            \]
          </div>

          <p>
            Formalmente, esto se expresa mediante las probabilidades de transición de un paso:
          </p>

          <div class="formula-box">
            <div class="label">Probabilidades de transición</div>
            \[
            \mathbb{P}(X_{n+1}=(x+1,y)\mid X_n=(x,y))=\frac{1}{4},
            \]
            \[
            \mathbb{P}(X_{n+1}=(x-1,y)\mid X_n=(x,y))=\frac{1}{4},
            \]
            \[
            \mathbb{P}(X_{n+1}=(x,y+1)\mid X_n=(x,y))=\frac{1}{4},
            \]
            \[
            \mathbb{P}(X_{n+1}=(x,y-1)\mid X_n=(x,y))=\frac{1}{4},
            \]
          </div>

          <p>
            y todas las demás probabilidades de transición en un solo paso son iguales a cero.
          </p>

          <p>
            Esta regla refleja la regularidad de grado 4 del grafo: como cada vértice tiene exactamente cuatro vecinos y no existe ninguna dirección privilegiada, el movimiento es completamente simétrico. En consecuencia, el paseo no presenta deriva ni tendencia sistemática hacia ninguna región del plano. Su evolución responde exclusivamente a elecciones aleatorias equiprobables entre las cuatro direcciones cardinales.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.3. Interpretación como cadena de Markov</span>
            <span class="topic-subtitle">La evolución futura depende solo del estado presente</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El paseo aleatorio simple sobre \( \mathbb{Z}^2 \) constituye una cadena de Markov homogénea sobre un espacio de estados infinito numerable. La propiedad de Markov significa que la distribución de la posición futura depende únicamente del estado presente y no de la historia previa del proceso. En términos formales,
          </p>

          <div class="formula-box">
            <div class="label">Propiedad de Markov</div>
            \[
            \mathbb{P}(X_{n+1}=z\mid X_n=x,\;X_{n-1}=x_{n-1},\ldots,X_0=x_0)
            =
            \mathbb{P}(X_{n+1}=z\mid X_n=x).
            \]
          </div>

          <p>
            Es decir, para predecir probabilísticamente el siguiente paso, basta conocer la posición actual del caminante; el camino recorrido antes de llegar allí no aporta información adicional relevante para la transición inmediata.
          </p>

          <p>
            Además, la cadena es homogénea en el tiempo porque la regla de transición no cambia con \(n\). Las probabilidades de moverse desde un vértice a cualquiera de sus vecinos son siempre las mismas, independientemente del instante en que se observe el proceso. Esto permite definir de manera natural las probabilidades de transición en \(n\) pasos mediante la notación
          </p>

          <div class="formula-box">
            <div class="label">Probabilidad de transición en n pasos</div>
            \[
            p_n(x,y)=\mathbb{P}(X_n=y\mid X_0=x).
            \]
          </div>

          <p>
            En particular, si se toma como estado inicial el origen, la expresión
          </p>

          <div class="formula-box">
            <div class="label">Probabilidad de retorno al origen</div>
            \[
            p_n(0,0)=\mathbb{P}(X_n=(0,0)\mid X_0=(0,0))
            \]
          </div>

          <p>
            representa la probabilidad de retorno al origen en el tiempo \(n\). Esta cantidad ocupa un lugar central en el estudio del proceso, ya que permite analizar la frecuencia, la estructura temporal y el comportamiento asintótico de los retornos.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.4. Paridad de los retornos</span>
            <span class="topic-subtitle">La bipartición del grafo impone retornos solo en tiempos pares</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Una propiedad fundamental del paseo aleatorio sobre \( \mathbb{Z}^2 \) proviene de la bipartición del grafo. Como los vértices pueden separarse según la paridad de la suma de sus coordenadas, cada paso del paseo cambia necesariamente dicha paridad. En efecto, si el caminante se encuentra en \((x,y)\), entonces al pasar a un vecino inmediato una de las coordenadas cambia en una unidad, por lo que la suma \(x+y\) cambia de paridad.
          </p>

          <p>
            Si el proceso comienza en el origen \((0,0)\), donde \(x+y=0\) es par, entonces después de un paso el caminante se encuentra en un vértice de paridad impar; después de dos pasos vuelve a un vértice de paridad par; y así sucesivamente. Como consecuencia, el retorno al origen solo puede ocurrir en tiempos pares.
          </p>

          <div class="formula-box">
            <div class="label">Restricción de paridad</div>
            \[
            p_{2n+1}(0,0)=0
            \qquad \text{para todo } n\ge 0.
            \]
          </div>

          <p>
            Es decir, la probabilidad de regresar al origen tras un número impar de pasos es exactamente nula. Esta restricción no depende de un cálculo particular, sino de la estructura bipartita del grafo. La paridad constituye así una limitación geométrica fundamental sobre las trayectorias posibles del paseo aleatorio.
          </p>

          <p>
            Esta observación tiene además una importancia técnica considerable, ya que muchas fórmulas asintóticas y resultados clásicos sobre probabilidades de retorno se expresan específicamente para tiempos pares, por ejemplo \(p_{2n}(0,0)\), que es la secuencia verdaderamente relevante en dimensión dos.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.5. Recurrencia en dimensión 2</span>
            <span class="topic-subtitle">El caminante vuelve al origen con probabilidad uno</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Uno de los resultados más clásicos y profundos de la teoría del paseo aleatorio afirma que el paseo aleatorio simple sobre \( \mathbb{Z}^2 \) es recurrente. Esto significa que, si el proceso comienza en el origen, entonces retorna a ese vértice con probabilidad igual a uno. Más aún, no solo vuelve una vez, sino que regresa infinitas veces con probabilidad uno.
          </p>

          <div class="formula-box">
            <div class="label">Recurrencia</div>
            \[
            \mathbb{P}(\text{el paseo retorna al origen alguna vez}\mid X_0=(0,0))=1.
            \]
          </div>

          <p>
            De manera más fuerte, el origen es visitado infinitas veces casi seguramente. Este hecho resulta notable porque el espacio es infinito: el caminante puede alejarse arbitrariamente del punto de partida y, sin embargo, en dimensión dos la aleatoriedad del movimiento no basta para evitar el retorno.
          </p>

          <p>
            La recurrencia distingue de manera esencial al caso bidimensional de otros contextos. En dimensión uno, el paseo también es recurrente, pero en dimensiones tres o superiores se vuelve transiente, es decir, existe una probabilidad positiva de no regresar nunca al punto inicial. La dimensión dos ocupa, por tanto, una posición crítica y especialmente interesante en la teoría.
          </p>

          <p>
            Desde el punto de vista analítico, la recurrencia está estrechamente vinculada con el comportamiento de las probabilidades de retorno \(p_{2n}(0,0)\). Aunque estas probabilidades tienden a cero cuando \(n\to\infty\), lo hacen suficientemente lentamente como para que la suma de todos los retornos posibles diverja. Esta divergencia expresa que, a largo plazo, el caminante sigue teniendo oportunidades acumuladas de volver una y otra vez al origen.
          </p>

          <p>
            Intuitivamente, el resultado indica que en la red bidimensional el paseo explora indefinidamente nuevas regiones, pero no se dispersa de manera irreversible. El equilibrio entre libertad de movimiento e insistencia en revisitar zonas ya recorridas es precisamente una de las características más distintivas de la dimensión dos.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.6. Significado probabilístico y estructural del modelo</span>
            <span class="topic-subtitle">Un modelo central para grafos, probabilidad y difusión discreta</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El paseo aleatorio simple sobre \( \mathbb{Z}^2 \) no debe verse únicamente como un ejemplo elemental, sino como un modelo fundamental para una amplia variedad de problemas matemáticos y físicos. Desde la perspectiva de la teoría de grafos, pone en juego la conectividad, la regularidad, la bipartición y las simetrías del grafo infinito. Desde la probabilidad, proporciona un caso central de cadena de Markov sobre un espacio de estados numerable e infinito.
          </p>

          <p>
            Además, el modelo sirve como punto de partida para estudiar tiempos de retorno, tiempos de visita, funciones de Green, núcleos de transición, comportamiento asintótico y fenómenos de difusión. También aparece en aproximaciones discretas de procesos continuos, como el movimiento browniano, y en modelos de transporte, física estadística y dinámica en medios homogéneos.
          </p>

          <p>
            Por ello, comprender la definición y las propiedades básicas del paseo aleatorio sobre \( \mathbb{Z}^2 \) es esencial para el desarrollo posterior de resultados más avanzados. La simplicidad de la regla local contrasta con la profundidad de sus consecuencias globales, lo que explica su lugar central en la matemática moderna.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">7.7. Síntesis de la sección</span>
            <span class="topic-subtitle">Resumen estructural del proceso y su relevancia matemática</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            En síntesis, el paseo aleatorio simple sobre \( \mathbb{Z}^2 \) es un proceso estocástico discreto que evoluciona sobre la malla cuadrada infinita moviéndose, en cada paso, a uno de los cuatro vecinos inmediatos con probabilidad uniforme. Este proceso define una cadena de Markov homogénea sobre un espacio de estados infinito numerable, hereda la bipartición del grafo y, por ello, solo puede retornar al origen en tiempos pares. Su propiedad más destacada en dimensión dos es la recurrencia: el caminante vuelve al origen con probabilidad uno e incluso lo visita infinitas veces casi seguramente.
          </p>

          <p>
            Estas características convierten al paseo aleatorio sobre \( \mathbb{Z}^2 \) en el marco natural para estudiar de manera rigurosa las probabilidades de retorno y los resultados asintóticos que se desarrollarán en las secciones siguientes.
          </p>
        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        El paseo aleatorio simple sobre \( \mathbb{Z}^2 \) es una cadena de Markov homogénea sobre una red bidimensional infinita, donde en cada paso el caminante elige uniformemente uno de los cuatro vecinos inmediatos del vértice actual.
      </p>
      <p>
        La bipartición del grafo impone que los retornos al origen solo puedan ocurrir en tiempos pares, y la estructura particular de la dimensión dos hace que el proceso sea recurrente: el caminante regresa al origen con probabilidad uno y, de hecho, lo visita infinitas veces casi seguramente.
      </p>
      <p>
        Esta combinación de simplicidad local y riqueza global convierte al modelo en una herramienta central para estudiar probabilidades de retorno, difusión discreta, funciones de Green y leyes asintóticas sobre redes infinitas.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_paseo_aleatorio_z2', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_paseo_aleatorio_z2') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
# @title Qué demuestra el código

from IPython.display import display, HTML

html_code = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Qué demuestra el código</title>

  <script src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML" async></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">

  <style>
    :root {
      --bg-primary: linear-gradient(135deg, #f0f9ff 0%, #e0f2fe 100%);
      --bg-secondary: rgba(255, 255, 255, 0.88);
      --bg-tertiary: rgba(248, 250, 252, 0.82);
      --text-primary: #0c4a6e;
      --text-secondary: #075985;
      --accent-primary: #0ea5e9;
      --accent-secondary: #38bdf8;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.85);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --shadow-soft: 0 8px 20px rgba(14, 165, 233, 0.08);
      --border-radius: 20px;
      --transition: all 0.35s cubic-bezier(0.25, 0.8, 0.25, 1);
      --quote-bg: rgba(14, 165, 233, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(14, 165, 233, 0.12) 0%, rgba(56, 189, 248, 0.08) 100%);
      --table-header: rgba(14, 165, 233, 0.12);
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0c4a6e 0%, #1e3a8a 100%);
      --bg-secondary: rgba(30, 58, 138, 0.86);
      --bg-tertiary: rgba(30, 64, 175, 0.72);
      --text-primary: #f0f9ff;
      --text-secondary: #dbeafe;
      --accent-primary: #38bdf8;
      --accent-secondary: #7dd3fc;
      --border-color: rgba(255, 255, 255, 0.16);
      --shadow-card: 0 18px 40px rgba(0, 0, 0, 0.22);
      --shadow-soft: 0 8px 20px rgba(125, 211, 252, 0.10);
      --quote-bg: rgba(255, 255, 255, 0.05);
      --concept-bg: linear-gradient(135deg, rgba(56, 189, 248, 0.16) 0%, rgba(125, 211, 252, 0.10) 100%);
      --table-header: rgba(255, 255, 255, 0.08);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
      scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', sans-serif;
      line-height: 1.85;
      background: var(--bg-primary);
      color: var(--text-primary);
      transition: var(--transition);
      min-height: 100vh;
      position: relative;
      overflow-x: hidden;
      padding: 0;
    }

    .particles {
      position: fixed;
      inset: 0;
      pointer-events: none;
      z-index: -1;
    }

    .particle {
      position: absolute;
      border-radius: 50%;
      animation: float 24s infinite linear;
      opacity: 0;
      background: rgba(255, 255, 255, 0.55);
    }

    @keyframes float {
      0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
      10%, 90% { opacity: 0.55; }
      100% { transform: translateY(-10vh) rotate(360deg); opacity: 0; }
    }

    .theme-toggle {
      position: fixed;
      top: 1.4rem;
      right: 1.4rem;
      width: 56px;
      height: 56px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      backdrop-filter: blur(15px);
      box-shadow: var(--shadow-card);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary);
      transition: var(--transition);
      z-index: 1000;
    }

    .theme-toggle:hover {
      transform: scale(1.08) rotate(180deg);
    }

    .container {
      max-width: 1100px;
      margin: 0 auto;
      padding: 2rem 1.2rem 3rem;
    }

    .hero {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 28px;
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 2.2rem 2rem;
      margin-bottom: 1.6rem;
    }

    .eyebrow {
      display: inline-flex;
      align-items: center;
      gap: 0.55rem;
      font-size: 0.92rem;
      font-weight: 700;
      letter-spacing: 0.02em;
      color: var(--accent-primary);
      margin-bottom: 0.9rem;
      text-transform: uppercase;
    }

    .main-title {
      font-size: clamp(2rem, 5vw, 3.5rem);
      line-height: 1.12;
      font-weight: 800;
      background: var(--accent-gradient);
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.85rem;
    }

    .subtitle {
      color: var(--text-secondary);
      font-size: 1.06rem;
      max-width: 940px;
    }

    .block {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-card);
      backdrop-filter: blur(18px);
      padding: 1.8rem 1.6rem;
      margin-bottom: 1.3rem;
    }

    .block h2 {
      font-size: 1.55rem;
      color: var(--text-primary);
      margin-bottom: 1rem;
      display: flex;
      align-items: center;
      gap: 0.7rem;
    }

    .block p, .topic-content p {
      color: var(--text-secondary);
      margin-bottom: 1rem;
      text-align: justify;
    }

    .accordion {
      display: flex;
      flex-direction: column;
      gap: 1rem;
      margin-top: 1.2rem;
    }

    .topic-card {
      background: var(--bg-secondary);
      border: 2px solid var(--border-color);
      border-radius: 20px;
      box-shadow: var(--shadow-card);
      overflow: hidden;
      transition: var(--transition);
      backdrop-filter: blur(18px);
    }

    .topic-header {
      cursor: pointer;
      padding: 1.3rem 1.4rem;
      display: flex;
      justify-content: space-between;
      align-items: center;
      gap: 1rem;
    }

    .topic-title {
      font-size: 1.18rem;
      font-weight: 700;
      color: var(--text-primary);
    }

    .topic-subtitle {
      display: block;
      font-size: 0.95rem;
      color: var(--text-secondary);
      font-weight: 400;
      margin-top: 0.25rem;
    }

    .expand-icon {
      font-size: 1.1rem;
      color: var(--text-secondary);
      transition: var(--transition);
      flex-shrink: 0;
    }

    .topic-card.open .expand-icon {
      transform: rotate(180deg);
    }

    .topic-content {
      max-height: 0;
      overflow: hidden;
      transition: max-height 1s ease, padding 1s ease;
      background: var(--bg-tertiary);
      padding: 0 1.4rem;
    }

    .topic-card.open .topic-content {
      max-height: 18000px;
      padding: 1.2rem 1.4rem 1.5rem 1.4rem;
      border-top: 1px solid var(--border-color);
    }

    .formula-box {
      background: var(--bg-secondary);
      border: 1px solid var(--border-color);
      border-left: 4px solid var(--accent-primary);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
      box-shadow: var(--shadow-soft);
      overflow-x: auto;
    }

    .formula-box .label {
      font-size: 0.88rem;
      font-weight: 700;
      color: var(--accent-primary);
      margin-bottom: 0.45rem;
      text-transform: uppercase;
      letter-spacing: 0.02em;
    }

    .example-box {
      background: var(--quote-bg);
      border: 1px solid var(--border-color);
      border-radius: 14px;
      padding: 1rem 1.1rem;
      margin: 1rem 0 1.1rem;
    }

    .example-box strong {
      color: var(--text-primary);
    }

    .concept-card {
      background: var(--concept-bg);
      border: 2px solid var(--border-color);
      border-radius: 22px;
      box-shadow: var(--shadow-card);
      padding: 1.6rem 1.5rem;
      margin-top: 1.4rem;
    }

    .concept-card h3 {
      margin: 0 0 0.7rem 0;
      padding: 0;
      border: none;
      font-size: 1.15rem;
      color: var(--text-primary);
      display: flex;
      align-items: center;
      gap: 0.65rem;
    }

    .concept-card p {
      margin: 0 0 0.9rem 0;
      font-size: 1rem;
      color: var(--text-secondary);
      text-align: justify;
    }

    .footer {
      text-align: center;
      margin-top: 2rem;
      padding-top: 1.4rem;
      border-top: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      opacity: 0.9;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem 0.8rem 2rem;
      }
      .hero, .block, .concept-card {
        padding: 1.25rem 1rem;
      }
      .main-title {
        font-size: 2rem;
      }
      .block h2 {
        font-size: 1.3rem;
      }
      .topic-header,
      .topic-card.open .topic-content {
        padding-left: 1rem;
        padding-right: 1rem;
      }
    }
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema">
    <i class="fas fa-moon" id="theme-icon"></i>
  </button>

  <div class="container">

    <section class="hero">
      <div class="eyebrow">
        <i class="fas fa-code"></i>
        Exploración numérica del paseo aleatorio
      </div>
      <h1 class="main-title">8. Qué demuestra el código</h1>
      <p class="subtitle">
        El código no constituye una demostración analítica en sentido estricto, sino una herramienta de exploración numérica del paseo aleatorio simple sobre \( \mathbb{Z}^2 \), capaz de producir evidencia computacional visible, cuantificable y reproducible.
      </p>
    </section>

    <section class="block">
      <h2><i class="fas fa-lightbulb"></i> Visión general</h2>
      <p>
        Su función principal es generar una gran cantidad de trayectorias aleatorias independientes, registrar en cuáles de ellas el caminante regresa al origen en tiempos determinados y, a partir de esas observaciones, construir estimaciones empíricas de probabilidades de retorno. De este modo, el código permite comparar el comportamiento observado en la simulación con las leyes asintóticas conocidas para el paseo aleatorio bidimensional.
      </p>
    </section>

    <section class="accordion">

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.1. Simulación empírica de muchos paseos</span>
            <span class="topic-subtitle">No una trayectoria aislada, sino una gran muestra estadística</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            La primera idea central del código es que no trabaja con una trayectoria aislada, sino con una gran colección de paseos aleatorios simulados de manera independiente. Cada una de estas trayectorias representa una posible realización del proceso estocástico definido sobre \( \mathbb{Z}^2 \), en el que el caminante, partiendo del origen u otro vértice inicial fijado, se desplaza en cada paso a uno de sus cuatro vecinos inmediatos con probabilidad uniforme.
          </p>

          <p>
            Al repetir este experimento un número muy grande de veces, el código obtiene una muestra estadística suficientemente amplia como para estimar frecuencias de ciertos eventos de interés. Entre esos eventos, el más importante es el retorno al origen en tiempos específicos, especialmente en tiempos pares, ya que por la bipartición del grafo los retornos en tiempos impares son imposibles.
          </p>

          <p>
            Desde el punto de vista metodológico, la simulación empírica cumple aquí el papel de aproximar, mediante datos observados, cantidades probabilísticas teóricas que no se calculan a partir de una sola trayectoria, sino del comportamiento colectivo de muchas realizaciones del proceso.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.2. Estimación de \( p_{2n}(0,0) \)</span>
            <span class="topic-subtitle">La probabilidad de retorno al origen aproximada por frecuencias relativas</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Una de las cantidades centrales en el estudio del paseo aleatorio simple sobre \( \mathbb{Z}^2 \) es la probabilidad de retorno al origen luego de \(2n\) pasos, que se denota por
          </p>

          <div class="formula-box">
            <div class="label">Probabilidad teórica de retorno</div>
            \[
            p_{2n}(0,0)=\mathbb{P}(X_{2n}=(0,0)\mid X_0=(0,0)).
            \]
          </div>

          <p>
            Esta magnitud expresa la probabilidad de que el caminante, habiendo comenzado en el origen, vuelva exactamente a ese punto tras \(2n\) movimientos.
          </p>

          <p>
            El código aproxima esta probabilidad por medio de una frecuencia relativa. Si se simulan \(N\) trayectorias independientes y en \(R_{2n}\) de ellas el caminante se encuentra en el origen al cabo de \(2n\) pasos, entonces la estimación empírica es
          </p>

          <div class="formula-box">
            <div class="label">Estimación empírica</div>
            \[
            \widehat{p}_{2n}(0,0)=\frac{R_{2n}}{N}.
            \]
          </div>

          <p>
            Esta expresión no es exacta, pero se espera que se aproxime a la probabilidad verdadera cuando el número de simulaciones es grande. En otras palabras, el código reemplaza una magnitud teórica por una proporción observada experimentalmente. Así, la teoría probabilística se traduce en una metodología computacional concreta: contar retornos y dividir por el número total de ensayos.
          </p>

          <p>
            Este procedimiento es típico en simulación de Monte Carlo, donde probabilidades abstractas se estiman a partir de frecuencias obtenidas en muestras aleatorias suficientemente numerosas.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.3. Observación de la ley \( p_{2n}(0,0)\sim \frac{1}{\pi n} \)</span>
            <span class="topic-subtitle">La simulación como evidencia de una tendencia asintótica</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            El objetivo matemático más importante de la simulación es poner en evidencia la ley asintótica clásica
          </p>

          <div class="formula-box">
            <div class="label">Ley asintótica clásica</div>
            \[
            p_{2n}(0,0)\sim \frac{1}{\pi n},
            \qquad n\to\infty.
            \]
          </div>

          <p>
            Esta relación significa que, para valores grandes de \(n\), la probabilidad de retorno en el tiempo \(2n\) se comporta aproximadamente como \(1/(\pi n)\). No se afirma que ambas cantidades sean iguales para todo \(n\), sino que su cociente tiende a 1 cuando \(n\) crece.
          </p>

          <p>
            El código busca mostrar empíricamente esta ley comparando los valores simulados de \(\widehat{p}_{2n}(0,0)\) con la expresión teórica \(1/(\pi n)\). Si la simulación está bien implementada y el número de trayectorias es suficientemente grande, debería observarse que ambas cantidades se aproximan progresivamente a medida que \(n\) aumenta, aunque siempre con ciertas fluctuaciones debidas al azar y al tamaño finito de la muestra.
          </p>

          <p>
            Por lo tanto, lo que el código pone de manifiesto no es una igualdad exacta, sino una tendencia asintótica. La simulación permite visualizar que la probabilidad de retorno decrece con el tiempo y que lo hace aproximadamente con el orden \(1/n\), con la constante precisa \(1/\pi\), lo cual constituye una de las firmas más características del paseo aleatorio simple en dimensión dos.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.4. Uso de sumas acumuladas para estimar \( \pi \)</span>
            <span class="topic-subtitle">Cómo estabilizar la señal y extraer una constante analítica</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Una idea especialmente interesante consiste en utilizar la ley asintótica de retorno para construir estimadores numéricos de \( \pi \). Dado que
          </p>

          <div class="formula-box">
            <div class="label">Punto de partida</div>
            \[
            p_{2n}(0,0)\sim \frac{1}{\pi n},
            \]
          </div>

          <p>
            puede reorganizarse informalmente esta expresión para obtener
          </p>

          <div class="formula-box">
            <div class="label">Estimador puntual informal</div>
            \[
            \pi \approx \frac{1}{n\,p_{2n}(0,0)}
            \]
          </div>

          <p>
            cuando \(n\) es grande. Sustituyendo la probabilidad exacta por su estimación empírica, se obtiene un estimador computacional de \( \pi \) basado en los retornos del paseo aleatorio.
          </p>

          <p>
            Sin embargo, las estimaciones puntuales de \(p_{2n}(0,0)\) suelen presentar fluctuaciones estadísticas apreciables, especialmente cuando el número de simulaciones no es extremadamente grande. Por esta razón, el código puede recurrir al uso de sumas acumuladas o promedios acumulados, que ayudan a suavizar la variabilidad inherente al muestreo aleatorio.
          </p>

          <p>
            La idea consiste en no basarse únicamente en un valor aislado de retorno, sino en la información agregada de muchos tiempos. Por ejemplo, si se considera la suma acumulada de probabilidades de retorno o de sus estimaciones,
          </p>

          <div class="formula-box">
            <div class="label">Suma acumulada</div>
            \[
            \sum_{k=1}^{n}\widehat{p}_{2k}(0,0),
            \]
          </div>

          <p>
            entonces esa suma presenta un comportamiento más estable que cada término individual. Dado que la serie armónica aparece en la suma de términos del tipo \(1/(\pi k)\), surge una relación aproximada de la forma
          </p>

          <div class="formula-box">
            <div class="label">Comportamiento acumulado</div>
            \[
            \sum_{k=1}^{n}p_{2k}(0,0)\approx \frac{1}{\pi}\ln n+\text{constante},
            \]
          </div>

          <p>
            lo que permite extraer nuevamente información sobre \( \pi \), ahora a partir de una magnitud acumulativa menos sensible a las oscilaciones aleatorias. En este sentido, el uso de sumas acumuladas no altera la teoría, sino que mejora la estabilidad numérica de la estimación y facilita la observación de la ley asintótica en datos simulados.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.5. Diferencia entre evidencia numérica y demostración formal</span>
            <span class="topic-subtitle">La simulación no prueba, pero sí hace visible la teoría</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Aunque el código proporciona resultados consistentes con la teoría, es fundamental distinguir entre evidencia computacional y demostración matemática. Una simulación, por extensa y precisa que sea, no constituye una prueba formal de una propiedad asintótica. Lo que hace es generar datos que apoyan, ilustran o verifican empíricamente una hipótesis teórica.
          </p>

          <p>
            En cambio, una demostración matemática exige argumentos rigurosos que establezcan la validez universal del resultado para todos los casos contemplados. En el problema del paseo aleatorio sobre \( \mathbb{Z}^2 \), demostrar que
          </p>

          <div class="formula-box">
            <div class="label">Resultado teórico a demostrar</div>
            \[
            p_{2n}(0,0)\sim \frac{1}{\pi n}
            \]
          </div>

          <p>
            requiere técnicas analíticas específicas, como el estudio de coeficientes combinatorios, funciones generadoras, transformadas de Fourier o resultados profundos de probabilidad sobre cadenas de Markov y paseos aleatorios.
          </p>

          <p>
            La simulación, por tanto, no reemplaza a la prueba, pero sí cumple un papel matemáticamente valioso. Permite verificar que la implementación computacional reproduce el comportamiento esperado, ofrece intuición sobre la magnitud y la velocidad del decaimiento de las probabilidades de retorno, y muestra de forma concreta cómo una constante tan fundamental como \( \pi \) puede emerger en un problema de caminatas aleatorias sobre una red discreta.
          </p>

          <p>
            En este sentido, el código actúa como puente entre teoría y experimento numérico: no demuestra la ley, pero la hace visible.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.6. Alcance matemático de la simulación</span>
            <span class="topic-subtitle">Lo que sí puede afirmarse a partir del experimento computacional</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            Desde una perspectiva más amplia, el código demuestra que un fenómeno probabilístico teórico puede observarse computacionalmente con notable fidelidad. A través de un gran número de trayectorias simuladas, se pone en evidencia que las probabilidades de retorno al origen no se comportan de manera arbitraria, sino que siguen una regularidad asintótica precisa. Esto refuerza la idea de que el paseo aleatorio, pese a su definición local y aleatoria, presenta leyes globales bien estructuradas.
          </p>

          <p>
            Además, la simulación muestra que el estudio empírico no solo sirve para confirmar cualitativamente una teoría, sino también para estimar cantidades concretas. En particular, permite aproximar probabilidades de retorno y, a partir de ellas, construir estimadores numéricos de constantes analíticas. El hecho de que \( \pi \) surja en este contexto ilustra de manera especialmente clara la profundidad del vínculo entre combinatoria, probabilidad y análisis.
          </p>
        </div>
      </div>

      <div class="topic-card">
        <div class="topic-header">
          <div>
            <span class="topic-title">8.7. Síntesis de la sección</span>
            <span class="topic-subtitle">Resumen conceptual del valor matemático del código</span>
          </div>
          <i class="fas fa-chevron-down expand-icon"></i>
        </div>
        <div class="topic-content">
          <p>
            En síntesis, el código no demuestra de forma teórica los resultados del paseo aleatorio sobre \( \mathbb{Z}^2 \), sino que realiza una simulación numérica de muchas trayectorias independientes para estimar frecuencias de retorno al origen. A partir de esas frecuencias, aproxima la probabilidad \(p_{2n}(0,0)\), observa empíricamente la ley asintótica
          </p>

          <div class="formula-box">
            <div class="label">Ley observada empíricamente</div>
            \[
            p_{2n}(0,0)\sim \frac{1}{\pi n},
            \]
          </div>

          <p>
            y utiliza sumas acumuladas para construir estimadores más estables de \( \pi \). Su importancia reside en que ofrece evidencia computacional coherente con la teoría y permite visualizar, de manera concreta, cómo una ley probabilística profunda puede emerger del comportamiento agregado de miles o millones de trayectorias aleatorias.
          </p>
        </div>
      </div>

    </section>

    <section class="concept-card">
      <h3><i class="fas fa-key"></i> Key Concept</h3>
      <p>
        El código no prueba analíticamente la teoría del paseo aleatorio sobre \( \mathbb{Z}^2 \), pero sí proporciona evidencia computacional clara y reproducible de sus leyes más importantes.
      </p>
      <p>
        Mediante simulación de Monte Carlo, estima probabilidades de retorno al origen, muestra empíricamente el comportamiento asintótico \(p_{2n}(0,0)\sim 1/(\pi n)\) y permite construir aproximaciones numéricas de \( \pi \) a partir de los datos generados.
      </p>
      <p>
        Su valor reside en tender un puente entre formulación abstracta y observación concreta: no reemplaza la demostración, pero vuelve visible la estructura probabilística profunda del modelo.
      </p>
    </section>

    <div class="footer">
      Material elaborado por el profesor Sergio Gevatschnaider.
    </div>
  </div>

  <script>
    (function () {
      const themeToggleButton = document.getElementById('themeToggleButton');
      const themeIcon = document.getElementById('theme-icon');
      const bodyEl = document.body;

      function setTheme(theme) {
        bodyEl.setAttribute('data-theme', theme);
        localStorage.setItem('theme_que_demuestra_el_codigo', theme);
        if (themeIcon) {
          themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }
      }

      if (themeToggleButton) {
        themeToggleButton.addEventListener('click', () => {
          const current = bodyEl.getAttribute('data-theme') || 'dark';
          setTheme(current === 'dark' ? 'light' : 'dark');
        });
      }

      const savedTheme = localStorage.getItem('theme_que_demuestra_el_codigo') || 'dark';
      setTheme(savedTheme);

      const container = document.getElementById('particles-container');
      if (container) {
        for (let i = 0; i < 28; i++) {
          const p = document.createElement('div');
          p.className = 'particle';
          p.style.left = Math.random() * 100 + 'vw';
          const size = Math.random() * 5 + 2;
          p.style.width = size + 'px';
          p.style.height = size + 'px';
          p.style.animationDelay = Math.random() * -20 + 's';
          p.style.animationDuration = (16 + Math.random() * 12) + 's';
          container.appendChild(p);
        }
      }

      document.querySelectorAll('.topic-header').forEach(header => {
        header.addEventListener('click', () => {
          const card = header.parentElement;
          card.classList.toggle('open');
          if (window.MathJax && window.MathJax.Hub) {
            setTimeout(() => {
              MathJax.Hub.Queue(["Typeset", MathJax.Hub]);
            }, 250);
          }
        });
      });

      const firstCard = document.querySelector('.topic-card');
      if (firstCard) {
        firstCard.classList.add('open');
      }
    })();
  </script>
</body>
</html>
"""

display(HTML(html_code))
```
