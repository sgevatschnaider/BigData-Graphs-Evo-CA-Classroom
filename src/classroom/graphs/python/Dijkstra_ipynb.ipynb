```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>El Algoritmo de Dijkstra</title>
    <style>
        /* Estilos básicos */
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
            background-blend-mode: lighten;
        }
        .navbar {
            background-color: #0066cc;
            color: white;
            padding: 10px 15px;
            text-align: center;
            font-size: 1.2em;
            position: fixed;
            width: 100%;
            top: 0;
            left: 0;
            z-index: 1000;
        }
        body {
            padding-top: 60px;
        }
        h1, h2, h3 {
            color: #333;
            text-align: center;
        }
        h1 { font-size: 2.5em; }
        h2 { color: #0066cc; font-size: 2em; }
        h3 { font-size: 1.5em; }
        p { font-size: 1.2em; margin-bottom: 15px; }
        ul, ol { margin-left: 20px; margin-bottom: 20px; }
        .card {
            background-color: #ffffff;
            border-radius: 8px;
            box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);
            padding: 20px;
            margin-bottom: 20px;
        }
        .divider {
            height: 2px;
            background-color: #e0e0e0;
            margin: 30px 0;
            border-radius: 4px;
        }
        .collapsible {
            cursor: pointer;
            font-weight: bold;
            text-align: left;
            background-color: #0066cc;
            color: white;
            border: none;
            outline: none;
            font-size: 1.1em;
            padding: 10px;
            width: 100%;
            text-align: left;
        }
        .content {
            display: none;
            padding: 15px;
            background-color: #f1f1f1;
        }
    </style>
</head>
<body>
    <div class="navbar">El Algoritmo de Dijkstra</div>

    <h1>El Algoritmo de Dijkstra</h1>

    <button class="collapsible">Introducción a Dijkstra y su Objetivo</button>
    <div class="content">
        <div class="card">
            <p>El <strong>Algoritmo de Dijkstra</strong> es un método ampliamente utilizado en teoría de grafos para encontrar el <strong>camino más corto</strong> desde un nodo de origen a todos los demás nodos en un grafo ponderado. Este algoritmo, propuesto por el científico informático Edsger Dijkstra en 1956, busca resolver problemas de optimización en redes, como la determinación de la ruta más eficiente en mapas, redes de transporte, y redes de comunicación.</p>
            <p>El objetivo principal de Dijkstra es encontrar el <strong>camino de menor costo</strong> desde un punto de partida hasta todos los puntos alcanzables en un grafo, bajo la condición de que las aristas tengan pesos no negativos. Esto lo hace especialmente útil en aplicaciones de redes de carreteras y sistemas de navegación, donde el peso suele representar la distancia, el tiempo de viaje o el costo asociado.</p>
        </div>
    </div>

    <div class="divider"></div>

    <button class="collapsible">Funcionamiento General del Algoritmo</button>
    <div class="content">
        <div class="card">
            <p>El algoritmo de Dijkstra sigue una estrategia <strong>voraz (greedy)</strong>, expandiendo el conjunto de nodos alcanzables en orden creciente de distancia desde el nodo de origen. A continuación se describen los pasos principales:</p>
            <ol>
                <li><strong>Inicialización</strong>: Se asigna una distancia de 0 al nodo de origen y una distancia infinita a todos los demás nodos. Esto significa que, al inicio, solo se conoce la distancia del nodo de origen a sí mismo, mientras que las distancias a los demás nodos son desconocidas y se representan como infinitas.</li>
                <li><strong>Expansión y actualización de distancias</strong>:
                    <ul>
                        <li>Se selecciona el nodo no visitado con la menor distancia conocida desde el origen. Este nodo pasa a ser "visitado" y no será reconsiderado.</li>
                        <li>Desde este nodo, se evalúan sus vecinos: para cada vecino, se calcula una distancia potencial como la suma de la distancia al nodo actual más el peso de la arista que conecta al vecino.</li>
                        <li>Si esta distancia calculada es menor que la distancia previamente registrada para el vecino, se actualiza la distancia del vecino y se guarda el nodo actual como predecesor en el camino de menor costo hacia él.</li>
                    </ul>
                </li>
                <li><strong>Repetición</strong>: Este proceso de expansión y actualización se repite hasta que todos los nodos alcanzables hayan sido visitados y sus distancias mínimas desde el origen se hayan determinado. El resultado es un árbol de caminos mínimos que muestra la ruta más corta desde el nodo de origen a todos los demás nodos.</li>
            </ol>
        </div>
    </div>

    <div class="divider"></div>

    <button class="collapsible">Expansión del Árbol de Caminos Mínimos</button>
    <div class="content">
        <div class="card">
            <p>A medida que el algoritmo avanza, construye un "árbol de caminos mínimos", en el que cada nodo tiene un camino definido desde el nodo de origen con la menor distancia posible. En este contexto, el "árbol" se refiere al conjunto de caminos de menor costo que conectan el nodo de origen con cada nodo alcanzable, evitando ciclos y minimizando las distancias.</p>
            <p>Este árbol se expande de forma ordenada, nodo por nodo, en función de la distancia mínima alcanzable, lo cual garantiza que cuando un nodo es marcado como "visitado", ya se ha encontrado el camino más corto hacia él.</p>
        </div>
    </div>

    <div class="divider"></div>

    <button class="collapsible">Uso de Estructuras de Datos Tipo Heap para Eficiencia</button>
    <div class="content">
        <div class="card">
            <p>Para mejorar la eficiencia del algoritmo, especialmente en grafos grandes, se utilizan estructuras de datos como el <strong>montículo (heap)</strong>, específicamente un heap de min-prioridad. Esto permite seleccionar rápidamente el nodo no visitado con la menor distancia conocida, lo que es crucial para reducir el tiempo de ejecución.</p>
            <p>En lugar de buscar el nodo de menor distancia en una lista de todos los nodos no visitados (lo cual sería costoso en tiempo de procesamiento), el heap de min-prioridad organiza los nodos de manera que el acceso al nodo de menor costo sea inmediato. Cuando se actualizan las distancias de los vecinos, el heap se ajusta dinámicamente para que la selección del siguiente nodo sea eficiente. Este enfoque reduce la complejidad temporal del algoritmo de O(V^2) (donde V es el número de nodos) a O((V + E) log V), donde E es el número de aristas del grafo.</p>
        </div>
    </div>

    <div class="divider"></div>

    <button class="collapsible">Ventajas y Limitaciones de Dijkstra</button>
    <div class="content">
        <div class="card">
            <ul>
                <li><strong>No admite pesos negativos</strong>: El algoritmo de Dijkstra no puede manejar grafos con aristas de peso negativo, ya que esto podría llevar a soluciones incorrectas. Los pesos negativos pueden causar que los caminos más cortos se vuelvan más largos después de una actualización, lo cual contradice la suposición fundamental del algoritmo.</li>
                <li><strong>Es voraz y no óptimo para todos los problemas de caminos mínimos</strong>: Aunque Dijkstra es ideal para grafos con pesos no negativos, otros algoritmos, como el algoritmo de Bellman-Ford, son más adecuados para grafos que contienen pesos negativos, aunque a costa de una mayor complejidad temporal.</li>
            </ul>
        </div>
    </div>

    <div class="divider"></div>

    <button class="collapsible">Conclusión</button>
    <div class="content">
        <div class="card">
            <p>El Algoritmo de Dijkstra es una herramienta poderosa y eficiente para encontrar caminos mínimos en redes con pesos no negativos. Su uso de una estrategia voraz y de estructuras de datos avanzadas como el heap de min-prioridad lo hace ideal para aplicaciones prácticas en redes de transporte y telecomunicaciones. Aunque tiene limitaciones en grafos con pesos negativos, sigue siendo uno de los algoritmos más importantes en teoría de grafos y optimización de rutas.</p>
        </div>
    </div>

    <script>
        var coll = document.getElementsByClassName("collapsible");
        for (var i = 0; i < coll.length; i++) {
            coll[i].addEventListener("click", function() {
                this.classList.toggle("active");
                var content = this.nextElementSibling;
                if (content.style.display === "block") {
                    content.style.display = "none";
                } else {
                    content.style.display = "block";
                }
            });
        }
    </script>
</body>
</html>
"""

# Mostrar el contenido HTML
display(HTML(html_content))


```
